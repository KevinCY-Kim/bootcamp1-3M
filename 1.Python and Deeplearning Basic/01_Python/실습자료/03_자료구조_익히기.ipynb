```python
# 정렬
fitness = ['스쿼트', '데드리프트', '벤치프레스', '바벨로우']

fit_sort = sorted(fitness)

print('origin:', fitness)
print('sorted:', fit_sort) # 가나다순

fit_sort2 = sorted(fitness, key=len)
print('sorted_by_len:', fit_sort2)
```

```python
fitness = ['스쿼트', '데드리프트', '벤치프레스', '바벨로우']

fit_re = sorted(fitness, reverse=True)

print('origin:', fitness)
print('reversed:', fit_re) # 가나다 역순
```

### 집합 메서드

1) 추가하기  
- add : 1개  
- update : 여러개

2) 삭제하기  
- remove : 1개/없는 걸 없애라고하면 key error  
- discard : 1개/없는 걸 없애라고해도 에러 X

```python
# 원소 하나 추가
comp = {'네이버', '카카오', '쿠팡', '배민'}

comp.add('당근')
comp
```

```python
# 원소 여러개 추가
comp.update({'라인', '토스'})
comp
```

```python
# 존재하는 원소만 제거
comp.remove('라인')
comp
```

```python
comp.remove('라인') #KeyError: '롯데'
comp
```

```python
# 명시하는 대상을 제거 : 존재하지 않아도 수행
comp.discard('라인')
comp
```

```python
movie1 = {'명량', '극한직업', '신과 함께', '기생충'}
movie2 = {'어벤져스', '겨울왕국', '아바타', '기생충'}
```

```python
#합집합
movie = movie1 | movie2
print(movie)
```

```python
movie = movie1.union(movie2)
print(movie)
```

```python
#교집합
movie = movie1 & movie2
print(movie)
```

```python
movie = movie1.intersection(movie2)
print(movie)
```

```python
#차집합
movie = movie1 - movie2
print(movie)
```

```python
movie = movie1.difference(movie2)
print(movie)
```

```python
#대칭차집합
movie = movie1^movie2
print(movie)
```

```python
movie = movie1.symmetric_difference(movie2)
print(movie)
```

```python
# movie1 에 덮어쓰기
# 합집합
movie1 = {'명량', '극한직업', '신과 함께', '기생충'}
movie2 = {'어벤져스', '겨울왕국', '아바타', '기생충'}
movie1.update(movie2)
movie1
```

```python
# 교집합
movie1 = {'명량', '극한직업', '신과 함께', '기생충'}
movie2 = {'어벤져스', '겨울왕국', '아바타', '기생충'}
movie1.intersection_update(movie2)
movie1
```

```python
# 차집합
movie1 = {'명량', '극한직업', '신과 함께', '기생충'}
movie2 = {'어벤져스', '겨울왕국', '아바타', '기생충'}
movie1.difference_update(movie2)
movie1
```

```python
# 대칭차집합
movie1 = {'명량', '극한직업', '신과 함께', '기생충'}
movie2 = {'어벤져스', '겨울왕국', '아바타', '기생충'}
movie1.symmetric_difference_update(movie2)
movie1
```

### 딕셔너리 in 용법

앞에서 감을 잡았을지 모르겠지만, 딕셔너리에서는 Value가 아니라, Key가 메인이다.  
'학교'라는 단어의 사전적 의미가 궁금하다고 가정해보자.  
그럼 사전에서 '학교'라는 단어를 검색하지 않겠는가.  
여기서 '학교' = key , '학교'에 대한 사전적 설명 = value 에 해당한다.  
  
즉, 딕셔너리에서 in 용법을 쓸 때는 당연하게도(?) key값으로 검색해야 한다.

```python
per_info = {'이름': '알디노',
            '직업': ['개발자', '강사'],
            '전공분야' : '머신러닝',
            '취미': ['볼링', '탁구', '수영'],
            '고향': '대전'}
per_info
```

```python
# 특정 키값이 있는지 확인 : 가능
'이름' in per_info
```

```python
'특기' in per_info
```

```python
'취미' in per_info
```

```python
# 특정 밸류가 있는지 확인 : 불가능
'알디노' in per_info # 이유는 in 용법으로는 키값 조회만 가능
```

```python
# value에서 특정 값이 있는지 찾고 싶을 때
'알디노' in per_info.values()
```
