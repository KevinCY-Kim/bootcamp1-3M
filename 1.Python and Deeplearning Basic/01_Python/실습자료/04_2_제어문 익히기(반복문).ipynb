### enumerate

### zip
이번에는 아래와 같이 두 개 이상의 리스트를 함께 순회하면서 동일시행을 반복하도록 코드를 작성해보자.  

[출력 결과]

샤브샤브: 20000원  
방어: 90000원  
호떡: 2000원

for 원소1, 원소2 in zip(리스트1, 리스트2)

```python
foods = ['샤브샤브', '방어', '호떡']
prices = [20000, 90000, 2000]

for food, price in zip(foods, prices):
    print(f'{food}: {price}원')
```

#### 문제 3. 아래 결과와 같이 두개의 리스트와 인덱스까지 함께 순회하면서 동일시행을 반복하도록 작성해보세요.  
[출력 결과]

메뉴1 : 볶음밥 - 아메리카노  
메뉴2 : 비빔밥 - 딸기라떼  
메뉴3 : 돈가스 - 버블티

```python
foods = ['볶음밥', '비빔밥', '돈가스']
drinks = ['아메리카노', '딸기라떼', '버블티']
for idx, (food, drink) in zip(foods, drinks):
    print(f"메뉴{idx+1} : {food} - {drink}")
```

```python
# @title 정답
foods = ['볶음밥', '비빔밥', '돈가스']
drinks = ['아메리카노', '딸기라떼', '버블티']

for idx, (food, drink) in enumerate(zip(foods, drinks)):
    print(f'메뉴{idx+1} : {food} - {drink}')
```

#### for i in range(시작점, 끝점+1, 변화간격) 용법

**굵은 텍스트**- i 는 변수다. (i am variables)
- 시작점은 안쓰면 0, 변화간격은 안쓰면 +1
- 끝점+1은 반드시 적어줘야함.

```python
for _ in range(5): # 5 --> 끝점+1 == 개수
    print('안녕하세요')
```

```python
for i in range(10): #시작점, 변화간격 안씀 -> 0부터 9까지 +1씩 변화
    print(i)
```

### break 용법
반복문을 중단하고 싶을 때,


break는 용도 자체가
"특정 조건에 해당"하면 반복을 중단하는 것이기 때문에  
조건문과 함께 사용되는 것이 일반적이다.

```python
# i 를 찾습니다!
for i in range(100):
    if i == 30:
        print(f'찾았습니다 : {i}')
        break
```

```python
for i in range(100):
    break
```

### continue 용법
해당 루프만 중단하고 싶을 때,  
바로 다음루프로 진행시키고 싶을 때


continue는 break의 반대??  
- break의 반대는 break를 안쓰는 것
- 반복문 중단의 반대는 반복문 진행이니까

continue도 중단이다.
- 단, 해당 루프만 중단하고, "다음루프로 넘어가"게 하는 문법이다.
- to be continue : 이번 화 끝! 다음화로 진행  

```python
# 30의 배수 찾기
for i in range(1, 100):
    if i%30 == 0:
        print(f'찾았습니다 : {i}')
        continue # continue 실행 -> 그럼 아래 코드는 실행X
    print(f'{i} 는 30의 배수가 아닙니다')
```

### 중첩 for문 용법
리딩스킬  
안쪽 for문부터 읽기

```python
for i in range(10):
    for j in range(5):
        print(i, j) # 리딩스킬 : 5바퀴 씩 10바퀴 = 총 50회 수행
```

3중 중첩 이상은 최대한 회피한다.  
시간 복잡도가 너무 높아진다.

```python
for i in range(100):
    for j in range(100):
        for k in range(100):
            print(i, j, k) # 100 * 100 * 100 = 100만회 실행
```

### while(조건) : 조건이 참인 동안 loop
특정 조건이 참인 동안 계속 반복하는 코드

```python
# 자기 자신에게 사칙연산할 때
a = 0

a += 1 # a = a + 1
print(a)
```

```python
cnt = 1

while cnt < 10: # 참이라면!
    print(cnt)
    cnt += 1  # cnt = cnt + 1 --> 참인 조건이 거짓이 되게 만들어주는 코드
```

### 무한루프 : While 1

```python
while 1: # while 조건 : 조건의 return True 라면 종속절 실행 -> loop
    print('hello')
    break
```

- 무한반복 코드를 쓸 때, while문을 자주 쓴다.  
- 무한반복?? 쓸 일이 있을까?

##### 무한반복문과 break를 이용
- 무한반복문은 진짜 무한반복할려고 쓴게 아니다.
- 편의상, 반복문을 중단하고자 하는 조건을 명시하고, 해당 조건을 충족할때까지 계속 반복시행하는 코드를 작성하고자 할 때
- 내가 위와 같이 생각하는게 더 편할 때
- 한정적 무한반복을 시행하고자 할 때

```python
#무한반복
count = 0
while 1:
    count += 1 #-> count = count + 1
    if count >= 10:
        print('loop finished')
        break #바로 상위의 loop 1개 종료
    print(count)

# 물론, while 의 조건으로 루프 시행 조건을 명시하는 것과 동일한 알고리즘이나,
# 코드 작성 입장에서 루프 중단 조건을 명시하고, break쓰는 것이 더 수월
```
