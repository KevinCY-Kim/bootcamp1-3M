```python
a = 5
b = 6
print(f'{a}는 {b}보다 크다')
```

## If A: 만약에 A라면

```python
a, b = 5, 6
if a > b: # :으로 끝나는 코드는 반드시 1개 이상의 종속 코드를 가져야 한다.
    print(f'{a}는 {b}보다 크다')
```

## else : 그렇지 않다면
생각2. 특정 조건을 만족하지 못할 경우에만 수행되어야 하는 코드는?


```python
a, b = 3, 3
if a > b:
    print('a는 b보다 크다')
else:
    print('a는 b보다 작다')
```

## elif B: 그렇지않고 B라면
생각3. 특정 조건을 만족하지 못하지만, 다른 조건은 만족하는 경우에는?


```python
a, b = 6, 7
if a > b:
    print('a는 b보다 크다') # 1
elif a < b:
    print('a는 b보다 작다') # 2
else:
    print('a는 b와 같다') # 3
```

## if 문 리딩스킬 : 참라면!
리딩스킬의 중요성 : 참이라면으로 읽으면 코드 응용력이 생긴다!
- 이게 컴퓨터 입장에서의 리딩이다.
- 알고리즘을 내부적으로 처리하는 방식 : "참/거짓"판단후 → 실행여부 결정
- 응용 : 앞으로 모든 코드를 볼 때, "컴퓨터는 어떻게 처리할까?" 고민하는 습관!

```python
# 아래 코드를 읽어보세요.
a = 5
if a > 0:
    print('a는 양수다')
```

대부분 "a가 0보다 크다면"이라고 읽을 것이다. 그러나 이렇게 읽지 않기를 권한다.  
"a > 0 가 참이라면" 이라고 읽자.  
- (조건문)이 참이라면 (종속코드 실행)
- (조건문 결과값)이 참이라면 (종속코드 실행)

if ( ) : ( ) 자리에는 참/거짓으로 판별될 수 있는 대상이라면 올 수 있다!!!!!

```python
# 응용 사례 1 : 리스트 값이 있을 때 원소 출력
lst = ['사과']

if lst: # lst 값이 있다면 == 참이라면
    print(lst)
else:
    print('빈리스트입니다.')

# 응용 사례 2 : 홀수/짝수 판별하여 출력

num = 3
if num%2: # 나머지가 참이라면??? == 나머지가 1이라면
    print('홀수 입니다.')
else:
    print('짝수 입니다.')
```

```python
# 리스트에 원소가 존재하는 경우에는 끝에 원소를 뽑아내자
a = ['사과', '딸기']
b = []

print(bool(a), bool(b)) # 값이 존재하면 참, 존재하지 않으면 거짓

# 리스트에 원소가 존재하는 경우
if a: # a의 결과값이 논리값으로 참인지 거짓인지 판별
    a.pop()
    print(a)

if b:
    b.pop()
    print(b)
else:
    print('원소가 없습니다')
```

### elif B: 그렇지않고 B라면
생각3. 특정 조건을 만족하지 못하지만, 다른 조건은 만족하는 경우에는?


가장 보고 싶은 영화의 잔여석을 먼저 확인해서 잔여석이 있으면 티켓요금을 안내해주고,  
그렇지않고, 두번째 보고 싶은 영화의 잔여석은 있다면 두번째 영화의 티켓요금을 안내해주고,
그렇지도 않다면 죄송하다는 문구를 띄워주도록 코드를 작성해보자

```python
movie_info = {'기생충' : {'잔여석' : 28,
                        '티켓요금' : 11000},
              '밀정' :  {'잔여석' : 0,
                        '티켓요금' : 11000}}
```

```python
my_movie = ['밀정', '기생충']
```

```python
seat = [movie_info[my_movie[0]]['잔여석'], movie_info[my_movie[1]]['잔여석']]
fare = [movie_info[my_movie[0]]['티켓요금'], movie_info[my_movie[1]]['티켓요금']]

if seat[0] != 0: #if (조건): = (조건)이 참이라면
  print(my_movie[0], '예매 도와드리겠습니다.')
  print('티켓요금은', fare[0], '입니다.')
elif seat[1] != 0:
  print('죄송합니다.')
  print(my_movie[0], '잔여석이 없습니다.')
  print('대신,', my_movie[1], '예매 도와드리겠습니다.')
  print('티켓요금은', fare[1], '입니다.')
else:
  print('죄송합니다.')
  print(my_movie[0], my_movie[1], '모두 잔여석이 없습니다.')
```

먼저, 보고싶은 영화가 원하는 시간대에 있는지 확인하고, 있다면 잔여석을 확인해주는 코드를 작성해보세요.

### if 안에 if
생각4.
특정 조건을 만족한 경우에, 또 다른 조건을 확인하고자 한다면?

if A:  
-> if B:

```python
# 둘의 차이 비교
# if A and B:

# if A:
#     if B:
```

```python
movie_info = {'기생충' :  {
                          '티켓요금' : 11000,
                          '7:00' : 28,
                          '10:00' : 0
                        },
              '밀정' :  {
                          '티켓요금' : 11000,
                          '7:30' : 23,
                          '10:30' : 102
                        }
              }
```

```python
# movie_info['기생충'] 의 자료구조는? -> 딕셔너리
# A in 딕셔너리 -> 딕셔너리의 키값에 A가 있다(있냐?) -> True or False
'7:00' in movie_info['기생충'] # keys = 티켓요금, 7:00, 10:00
```

```python
my_movie = '기생충'
my_time = '7:00'

if my_time in movie_info[my_movie]:
    print(my_movie, my_time, '영화 있습니다.')
    if movie_info[my_movie][my_time]: # if 잔여석:
        print(my_movie, '예매 도와드리겠습니다.')
        print('티켓요금은', movie_info[my_movie]['티켓요금'], '입니다.')
    else:
        print('죄송합니다.')
        print('잔여석이 없습니다.')
else:
    print('죄송합니다.')
    print(my_movie, '영화는', my_time, '시간대가 없습니다.')
```
