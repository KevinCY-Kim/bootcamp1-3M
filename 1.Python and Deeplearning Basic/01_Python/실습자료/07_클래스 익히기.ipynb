```python
class Unit(): # 이름을 지을 때, 대문자로 시작하기
    def __init__(self, hp=100, speed=100, atk=100): # 초기화함수(생성자) : 클래스 호출시 자동 동작
        self.hp = hp # adel.hp = 300
        self.speed = speed # adel.speed = 200
        self.atk = atk # adel.atk = 50

    def move(self, point):
        print(f'속력 {self.speed}으로 {point}까지 이동')

    def attack(self):
        print(f'상대에게 데미지 {self.atk}을 입힙니다.')
```

### 1, 아래 세부사항에 맞게 직접 클래스를 만들어봅시다.
- 클래스명 : User
- 인스턴스 변수 : number, username (default x)
- 클래스 변수 : count (default = 0)

```python
# @title 정답

class User():
    count = 0

    def __init__(self, number, username):
        self.number = number
        self.username = username

        User.count += 1
```

### 인스턴스 메서드, 클래스 메서드, 스태틱 메서드 차이 이해하기

어떤 애를 인스턴스 메서드로 만들고,  
어떤 애를 클래스 메서드로 만들건지 스스로 이해하는 것이 핵심이다.

```python
class Vehicle():
    print('클래스 변수 선언')
    instance_count = 0

    def __init__(self, wheel, color):
        self.wheel = wheel
        self.color = color
        Vehicle.instance_count += 1

    #인스턴스 메서드
    def move(self, speed):
        print('시속 {}km로 이동'.format(speed))

    #클래스 메서드
    @classmethod # decorator
    def count_instance(cls):
        print('현재 인스턴스 수 :{}'.format(Vehicle.instance_count))

    @staticmethod #스태틱 메서드
    def check_type(model_code):
        if model_code >= 20:
            print('전기차입니다.')
        else:
            print('디젤차입니다.')
```

```python
Vehicle.count_instance()
```

```python
my_vehicle = Vehicle(21, 'black')
```

```python
Vehicle.count_instance()
```

```python
model_code = 21
my_vehicle.check_type(model_code)
```

### 1, 아래 세부사항에 맞게 직접 클래스를 만들어봅시다.
- 위 1번에서 만든 User 클래스에 아래 세개의 메서드 추가하기
- 인스턴스 메서드 : display
    - return f'{self.number}번째 유저가 생성되었습니다.'
- 클래스 메서드 : get_count
    - return cls.count
- 스태틱 메서드 : welcome_message
    - return "Welcome to the User class!"

```python
# @title 정답

class User():
    count = 0

    def __init__(self, number, username):
        self.number = number
        self.username = username

        User.count += 1

    def display(self):
        return f'{self.number}번째 유저가 생성되었습니다.'

    @classmethod
    def get_count(cls):
        return cls.count

    @staticmethod
    def welcome_message():
        return "Welcome to the User class!"
```

### 2. 위 클래스를 호출하여 객체 1개를 만들고, 세개 함수를 모두 호출해봅시다.

```python
# @title 정답
user1 = User(1, 'paul')

print(user1.display())
print(User.get_count())
print(User.welcome_message())
```

## 상속

### 상속 개념 이해하기

새로운 클래스를 만들 때 기존의 클래스의 일부 요소를 그대로 가져다 쓰면 편하지 않을까?

```python
class Bicycle():
    def __init__(self, wheel=100, color='white'):
        self.wheel = wheel
        self.color = color

    def move(self, speed):
        print('시속 {}km로 이동'.format(speed))

    def turn(self, direction):
        print('{}방향으로 회전'.format(direction))

    def stop(self):
        print('정지')
```

- 부모클래스는 클래스명 오른쪽 괄호안에 적는다.
- 클래스 내에서 부모클래스명은 super()로 대신한다.
    - Bicycle.__ init __() → super(). __ init __()

```python
class ElectricBicycle(Bicycle):

    def __init__(self, wheel, color, battery):
        super().__init__(wheel, color)
        self.battery = battery

    def move(self, speed):
        super().move(speed)
        self.battery -= 1
        print(f'잔여 배터리 : {self.battery} %')

        if self.battery <= 0:
            super().stop()
```

```python
my_ev = ElectricBicycle(19, 'white', 100)
```

```python
my_ev.wheel, my_ev.color, my_ev.battery
```

```python
my_ev.move(10)
```

```python
my_ev.battery
```

### 1. 아래 세부사항에 맞게 직접 위 Bicycle 클래스를 상속받아 ElectricBicycle 클래스를 선언해봅시다.
- wheel, color 속성 가져오고, 추가로 electric_mode 속성(boolean) 추가하기
- electric_mode default값은 True
- move 메서드 선언하고, 부모클래스의 move 메서드 가져오고, 추가로 electric_mode에 따라 아래와 같이 출력 다르게 설정하기
    - if electric_mode == True :
        1. "electric mode로 동작합니다."
        2. '시속 {}km로 이동'
      
    - else :
        1. '수동 mode로 동작합니다.'
        2. '시속 {}km로 이동'


```python
class ElectricBicycle(Bicycle):

    def __init__(self, wheel, color, electric_mode == True):
        super().__init__(wheel, color)
        self.electric_mode == electric_mode

    def move(self, speed):
        if self.electric_mode:
            print(f'electric mode로 동작합니다')

        else:
            print(f'수동 mode로 동작합니다.')
```

```python
# @title 정답

class ElectricBicycle(Bicycle):

    def __init__(self, wheel, color, electric_mode=True):
        super().__init__(wheel, color)
        self.electric_mode = electric_mode

    def move(self, speed):
        if self.electric_mode:
            print('electric mode로 동작합니다.')
        else:
            print('수동 mode로 동작합니다.')
        super().move(speed)

    def stop(self):
        super().stop()
```
