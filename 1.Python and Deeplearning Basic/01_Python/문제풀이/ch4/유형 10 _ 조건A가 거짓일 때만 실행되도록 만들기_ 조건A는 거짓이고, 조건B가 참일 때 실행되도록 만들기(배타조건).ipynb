### 문제 23
string = '12aB34cD5'일 때, 주어진 문자열에서 알파벳 문자와 숫자를 처리하여 결과를 계산하는 프로그램을 작성해보세요.

* ord()함수는 문자를 유니코드 값으로 변환

1. 문자열을 순회하면서, 각 문자가 알파벳인지 숫자인지를 확인하고 만약 알파벳 문자라면, 해당 문자를 대문자로 변환한 후 ord() 함수를 사용하여 유니코드 값을 구합니다.
2. 구한 유니코드 값을 알파벳 문자들의 유니코드 값 합산에 더합니다.
3.  숫자인 경우에는 정수로 변환한 후 숫자들의 합에 더합니다.
4. 마지막으로 알파벳 문자들의 유니코드 값 합과 숫자들의 합을 곱한 결과를 계산하여 출력합니다.







````python
string = '12aB34cD5'
#ord()함수는 문자를 유니코드 값으로 변환

unicodes = 0
num = 0

for i in string:
    if i.isalpha():
        unicodes += ord(i.upper())
    elif i.isdigit():
        num += int(i)


result = unicodes * num
print(result)
````

### 정답

````python
string = '12aB34cD5'

alpha_sum = 0
num_sum = 0

for i in string:
    if i.isalpha():
        alpha_sum += ord(i.upper())
    elif i.isdigit():
        num_sum += int(i)

result = alpha_sum * num_sum

print(result)
````

### 문제 25
numbers = [10, 3, 8, 5, 2, 7]일 때, 주어진 숫자 리스트에서 두 번째로 작은 값과 두 번째로 큰 값을 찾는 프로그램을 작성해보세요.

1. 리스트의 첫 번째 원소를 초기값으로 설정합니다.
2. 리스트의 모든 원소를 순회하면서 두 번째로 작은 값과 두 번째로 큰 값을 찾습니다.
3. 초기값과 현재 원소를 비교하여 작은 값으면 두 번째 작은 값으로 설정, 큰 값이면 두 번째로 큰 값을 설정합니다.

````python
numbers = [10, 3, 8, 5, 2, 7]





#리스트의 첫번째 원소는 초기값
#주어진 숫자 리스트에서 두번째로 작은 값과 두번째로 큰값을 찾음
if numbers[0] < numbers[1]:
    min_num, min2 = numbers[0], numbers[1]
    max_num, max2 = numbers[1], numbers[0]
else:
    min_num, min2 = numbers[1], numbers[0]
    max_num, max2 = numbers[0], numbers[1]

#리스트의 모든 원소를 순회, 두번째로 작은 값과 두번째로 큰값을 찾음
#초기값과 현재 원소를 비교, 작은 값이면 두번째 작은 값으로 설정, 큰 값이면 두번쨰 큰 값으로 설정
for number in numbers[2:]:
    if number < min_num:
        min2, min_num = min_num, number
    elif number < min2:
        min2 = number

    if number > max_num:
        max2, max_num = max_num, number
    elif number > max2:
        max2 = number
print("두번째로 작은 값:", min2)
print("두번째로 큰 값:", max2)
````

### 정답

````python
numbers = [10, 3, 8, 5, 2, 7]

if numbers[0] < numbers[1]:
    min_num, min2 = numbers[0], numbers[1]
    max_num, max2 = numbers[1], numbers[0]
else:
    min_num, min2 = numbers[1], numbers[0]
    max_num, max2 = numbers[0], numbers[1]

for number in numbers[2:]:
    if number < min_num:
        min2, min_num = min_num, number
    elif number < min2:
        min2 = number

    if number > max_num:
        max2, max_num = max_num, number
    elif number > max2:
        max2 = number

print("두 번째로 작은 값:", min2)
print("두 번째로 큰 값:", max2)
````

- 그 외 경우, 해당 숫자에 3을 더한 값을 출력

### 문제 29
서울시의 세 개 구(강남구, 서초구, 송파구)에 대한 주택 데이터가 주어져 있습니다. 각 주택은 (지역명, 면적(㎡), 가격(원))의 형태로 표현됩니다. 이 데이터를 바탕으로, 각 지역별로 주택의 평균 1㎡당 가격을 계산하고, 특정 가격(예: 2,000,000원/㎡) 이상인 지역을 출력하는 프로그램을 작성하세요.
```
data = [
    ('강남구', 20, 100000000),
    ('서초구', 25, 120000000),
    ('송파구', 30, 150000000),
    ('강남구', 30, 130000000),
    ('서초구', 35, 160000000),
    ('송파구', 40, 180000000),
]
```

````python
data = [
    ('강남구', 20, 100000000),
    ('서초구', 25, 120000000),
    ('송파구', 30, 150000000),
    ('강남구', 30, 130000000),
    ('서초구', 35, 160000000),
    ('송파구', 40, 180000000),
]
# 각 지역별로 주택의 평균 1㎡당 가격을 계산하고, 특정 가격(예: 2,000,000원/㎡) 이상인 지역을 출력

area_prices = {}

for area, area_size, prices in data: # data를 튜플로 각각 저장
    price_per_sqm = prices / area_size # 가격을 평으로 나눠 평당 가격을 계산
    if area in area_prices: # 지역가격 딕셔너리에 지역이 있는지 확인
        area_prices[area].append(price_per_sqm) # 있다면 지역키값에 평당 가격을 업데이트
    else:
        area_prices[area] = [price_per_sqm] # 아직 등록되지 않은 지역이면 새리스트를 만들어 업데이트

for area, prices in area_prices.items(): # 위 area_prices 딕셔너리 내 저장된 지역과 평당가격을 묶어서 가져옴
    average_price = sum(prices) / len(prices) # 평당가격의 평균값을 계산, 합산 후 나눔
    area_prices[area] = average_price # 기존 가격리스트를 평균값으로 덮어씀, 딕셔너리의 리스트를 숫자(float)로 바꿈

threshold_price = 2000000 # 기준 가격 설정

for area, average_price in area_prices.items(): # 위 area_prices 반복
    if average_price >= threshold_price: # 기준 가격 이상만 소팅
     print(f"{area}의 평균 가격: {average_price:.2f}원 / 1평당") #출력 소수점 컷 2자리까지
````

### 정답

````python
data = [
    ('강남구', 20, 100000000),
    ('서초구', 25, 120000000),
    ('송파구', 30, 150000000),
    ('강남구', 30, 130000000),
    ('서초구', 35, 160000000),
    ('송파구', 40, 180000000),
]

area_prices = {}

for area, area_size, price in data:
    price_per_sqm = price / area_size
    if area in area_prices:
        area_prices[area].append(price_per_sqm)
    else:
        area_prices[area] = [price_per_sqm]

for area, prices in area_prices.items():
    average_price = sum(prices) / len(prices)
    area_prices[area] = average_price

threshold_price = 2000000

for area, average_price in area_prices.items():
    if average_price >= threshold_price:
        print(f"{area}의 평균 가격: {average_price:.2f}원 / 1평당")
````

### 문제 30
friends = ['예진', '은주', '보림', '희진', '종훈', '보승'] 다음은 마을 주민 리스트입니다. 뽀로로 마을에서 마피아 게임을 진행하려고 합니다.

맞출 수 있는 기회는 3회이며 '은주'가 마피아 일 때, for문을 사용해서 게임을 진행하는 프로그램을 작성하세요.
- input() 함수를 사용해서 기회가 몇 번 남았는지 출력하세요.
- 추리에 성공하면 "마피아를 발견했네요. 축하합니다."를 출력하세요.
- 추리에 실패하면 "다시 찾아보세요."를 출력하세요.
- 모든 추리 기회를 사용한 후에도 찾아내지 못한 경우, 게임을 종료하고 마피아의 이름을 출력하세요.

````python
friends = ['예진', '은주', '보림', '희진', '종훈', '보승']
mafia = '은주'
chance = 3

print("마피아 게임 참여자")
print(', '.join(friends))

for i in range(chance):
    remaining = chance - i
    guest = input(f"마피아는 누구일까요? 남은 기회는 {remaining}번 입니다.")
    if guest == mafia:
        print("마피아를 발견했네요. 축하합니다.")
        break
    elif i < chance - 1:
        print("다시 찾아보세요.")
    else:
        print(f"게임 종료. 마피아는 {mafia}였습니다.")
````
