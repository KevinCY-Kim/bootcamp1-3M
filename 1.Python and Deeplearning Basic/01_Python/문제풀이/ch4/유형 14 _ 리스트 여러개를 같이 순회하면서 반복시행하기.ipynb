### 문제 22
두 개의 리스트 num1과 num2가 주어졌을 때, 대응하는 위치에 대해
- num1의 원소가 짝수인 경우 num2의 해당 위치의 원소를,
- 홀수인 경우 num1의 해당 위치의 원소를 num_list에 추가하는 프로그램을 작성해보세요.
```
num1 = [1,2,3,4,5]
num2 = [2,4,6,8,10]
```

### 정답

````python
# num1와 num2 리스트의 각 요소를 확인하여 n1이 짝수면 n2 , 홀수면 n1을 num_list에 추가
num1 = [1,2,3,4,5]
num2 = [2,4,6,8,10]

num_list = []

for n1, n2 in zip(num1, num2):
    if n1 % 2 == 0:
        num_list.append(n2)
    else:
        num_list.append(n1)

print(num_list)
````

- "Country": 국가 이름
- "Capital": 수도

```
countries = ["USA", "Canada", "Japan"]

- 검거된 사건 수가 발생한 사건 수보다 많은 경우는 데이터에 오류가 있다고 가정하여 'Error'로 표시하세요.


```
자치구 = ['종로구', '중구', '용산구', '성동구', '관악구', '서초구', '강남구', '송파구', '강동구']
사건발생 = [4610, 5231, 3799, 0, 5328, 5542, 5069, 5698, 4014]
사건검거 = [5021, 3188, 2340, 0, 3810, 3750, 7304, 3799, 2955]

### 문제 30
당신은 로또 당첨 번호 맞추기 게임에 참가했습니다. 미리 정의된 당첨 번호가 있으며, 이 번호와 일치하는지 확인하는 게임입니다. 아래는 게임의 진행 방법입니다.

- 먼저, 미리 정의된 로또 당첨 번호 winning_numbers = [7, 14, 21, 28, 35, 42]가 있습니다.
- 사용자에게 1부터 45까지의 숫자 중 6개를 선택하도록 요청합니다. (공백으로 구분)
    - 사용자가 1 ~ 45 사이의 숫자를 입력하지 않거나, 6개의 숫자를 입력하지 않으면
    
    "잘못된 입력입니다. 1부터 45까지의 숫자 중 6개를 선택하세요."를 출력합니다.
- 사용자가 번호를 선택하면, 입력된 숫자와 당첨 번호를 비교하여 맞춘 번호의 개수를 확인하고 맞춘 개수를 출력합니다.
- 모든 번호를 맞춘 경우, 게임을 종료하고 "축하합니다! 모든 번호를 맞추셨습니다!"를 출력합니다.
- 하나의 번호라도 다른 경우, "아쉽지만 모든 번호를 맞추지 못했습니다. 다시 도전해보세요!"를 출력하고 게임을 계속 진행합니다.


출력 결과
```
로또 당첨 번호는 다음과 같습니다: [7, 14, 21, 28, 35, 42]
1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): 1 2 3 6 5
잘못된 입력입니다. 1부터 45까지의 숫자 중 6개를 선택하세요.
1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): 1 2 3 4 5 6
맞춘 번호 개수: 0
아쉽지만 모든 번호를 맞추지 못했습니다. 다시 도전해보세요!
1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): 7 14 21 28 35 42
맞춘 번호 개수: 6
축하합니다! 모든 번호를 맞추셨습니다!
```

### 정답

````python
winning_numbers = [7, 14, 21, 28, 35, 42]
print("로또 당첨 번호는 다음과 같습니다:", winning_numbers)

while True:
    user_input = input("1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): ")
    user_numbers = [int(num) for num in user_input.split()]

    if len(user_numbers) != 6 or any(num < 1 or num > 45 for num in user_numbers):
        print("잘못된 입력입니다. 1부터 45까지의 숫자 중 6개를 선택하세요.")
        continue

    matched_numbers = 0
    for number in user_numbers:
        if number in winning_numbers:
            matched_numbers += 1

    print("맞춘 번호 개수:", matched_numbers)

    if matched_numbers == 6:
        print("축하합니다! 모든 번호를 맞추셨습니다!")
        break
    else:
        print("아쉽지만 모든 번호를 맞추지 못했습니다. 다시 도전해보세요!")
````
