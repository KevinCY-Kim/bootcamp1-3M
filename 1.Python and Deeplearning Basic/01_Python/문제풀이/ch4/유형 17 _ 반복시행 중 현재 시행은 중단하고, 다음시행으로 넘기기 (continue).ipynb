### 문제 10
아래 코드를 읽고 출력 결과를 맞춰보세요

````python
keyinput = ["left", "right", "up", "right", "right"]
board = [11,11]

move = {
    'up': [0, 1],
    'down': [0, -1],
    'left': [-1, 0],
    'right': [1, 0],
}

n, m = (board[0]-1)//2, (board[1]-1)//2
x, y = 0, 0

for key in keyinput:
    dx, dy = move[key]
    nx, ny = x+dx, y+dy

    if nx not in range(-n, n+1):
        continue
    if ny not in range(-m, m+1):
        continue

    x, y = nx,ny

x, y
````

#### 출력 예시
```
---------- 긍정리뷰 ----------
깨끗하고 편안한 호텔입니다.
서비스가 좋았지만 시설은 조금 낡았어요.

```

````python
reviews = [
    ('깨끗하고 편안한 호텔입니다.', 4.5),
    ('서비스가 좋았지만 시설은 조금 낡았어요.', 3.8),
    ('위치도 좋고 방도 깨끗하고 아주 만족스러웠습니다.', 4.9),
    ('음식이 별로였고 시끄러워서 잠을 잘 못 잤어요.', 2.3),
    ('가격 대비 만족할만한 호텔입니다.', 4.2),
    ('친절한 직원들과 훌륭한 서비스를 경험했습니다.', 4.8)
]

positive_reviews = []
negative_reviews = []

positive_rating_sum = 0
negative_rating_sum = 0

positive_count = 0
negative_count = 0

for review, rating in reviews:
    if rating >= 3.5:
        positive_reviews.append(review)
        positive_rating_sum += rating
        positive_count += 1
        continue

    negative_reviews.append(review)
    negative_rating_sum += rating
    negative_count += 1

psitive_avg_rating = positive_rating_sum / positive_count if positive_count > 0 else 0
negative_avg_rating = negative_rating_sum / negative_count if negative_count > 0 else 0

print("---------- 긍정리뷰 ----------")
for review in positive_reviews:
    print(review)

print("긍정리뷰의 평균 평점:", positive_avg_rating)

print("---------- 부정리뷰 ----------")
for review in negative_reviews:
    print(review)

print("부정리뷰의 평균 평점:", negative_avg_rating)
````

### 문제 30
주어진 webtoon_author 리스트에서 평점이 가장 높은 상위 3개 웹툰을 top_recommended 리스트에 저장해보세요.

- top_recommended가 비어 있는 경우, 현재 웹툰을 top_recommended에 추가하고 다음 웹툰으로 이동합니다.
- top_recommended가 비어있지 않은 경우, top_recommended의 각 웹툰과 비교하여 평점이 높은 순서로 정렬합니다.
- 현재 웹툰의 평점이 이전 웹툰보다 높을 경우, 현재 웹툰을 top_recommended에 삽입합니다.
- top_recommended의 길이가 3을 초과하는 경우, 상위 3개 웹툰만 남기고 나머지는 제거합니다.

````python
webtoon_author = [
    {'author': 'SIU', 'rating': '9.93', 'title': 'Tower of God'},
    {'author': 'Chae Yong-taek / Hangaram', 'rating': '9.82', 'title': 'Beautyful Gangbari'},
    {'author': 'Seol-i / Yun Seong-won', 'rating': '9.95', 'title': 'Newbie Diary'},
    {'author': 'Jo Yong-seok', 'rating': '9.97', 'title': 'Jang Clan Escort...'},
    {'author': 'Jakka', 'rating': '8.84', 'title': 'Three Trials'}
]

recommended = []

for webtoon in webtoon_author:
    if float(webtoon['rating']) >= 9.9:
        recommended.append(webtoon['title'])

top_recommended = []

for webtoon in webtoon_author:
    if not top_recommended:
        top_recommended.append(webtoon)
        continue

    for top in top_recommended:
        if float(webtoon['rating']) >= float(top['rating']):
            top_recommended.insert(top_recommended.index(top), webtoon)
            if len(top_recommended) > 3:
                top_recommended = top_recommended[:3]
            break

top_recommended
````
