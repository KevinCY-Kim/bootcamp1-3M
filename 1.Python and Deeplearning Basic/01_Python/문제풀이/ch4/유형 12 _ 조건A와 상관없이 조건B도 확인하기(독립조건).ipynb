### 문제 24
주어진 좌표들을 이용하여 좌표(m,n)에 대해서 다음과 같은 규칙에 따라 계산을 수행하는 프로그램을 작성해보세요.

- 만약(m,n)이 (1,1)이라면, 결과는 0입니다.
- 만약(m,n)이 (1,n) 또는 (m,1)이라면, 결과는 m+n-2입니다.
- 그 외의 경우, 결과는 m-1 + (m*(n-1))입니다.
```
points = [(2,2), (-1,2), (3,-2),(-2,-3)]
```

### 정답

````python
points = [(2,2), (-1,2), (3,-2),(-2,-3)]

for point in points:
    m,n = point
    if m == 1 and n == 1:
        answer = 0
    if m == 1 or n == 1:
        answer = m + n - 2
    if m > n:
        temp = m
        m = n
        n = temp
answer = m-1 + (m*(n-1))
answer
````

- 맵기가 3이하인 경우, 일반떡볶이의 가격을 price 리스트에 추가합니다.
- 맵기가 3을 초과하는 경우, 일반떡복이의 가격에 500을 더한 값을 price 리스트에 추가합니다.


### 정답

````python
menu = {'일반떡볶이': {'맵기': [1, 2, 3, 4, 5], '가격': 5000}}

price = []

for spiciness in menu['일반떡볶이']['맵기']:
    if spiciness <= 3:
      price.append(menu['일반떡볶이']['가격'])
    if spiciness > 3:
      price.append(menu['일반떡볶이']['가격'] + 500)

print(price)
````

````python
movie_info = {
    '가디언즈 오브 갤럭시: Volume3': {'잔여석': 10, '티켓요금': 11000},
    '인어공주': {'잔여석': 7, '티켓요금': 11000},
    '분노의 질주: 라이드 오어 다이': {'잔여석': 3, '티켓요금': 11000},
    '범죄도시3': {'잔여석': 0, '티켓요금': 11000}
}
my_movie = [('범죄도시3', 3), ('분노의 질주: 라이드 오어 다이', 3), ('올빼미', 3)]

for movie, num_people in my_movie:
    if movie in movie_info:
        seat = movie_info[movie]['잔여석']
        fare = movie_info[movie]['티켓요금']

        if seat != 0:
            print(movie, '예매 도와드리겠습니다.')
            print('티켓요금은', fare, '입니다.')

            if num_people <= seat:
                total_fare = fare * num_people
                print('예매 인원 수:', num_people)
                print('총 결제금액은', total_fare, '입니다.')
                print("-" * 30)
            else:
                print('잔여석이 부족하여 예매할 수 없습니다.')
                print('현재 잔여석:', seat)
                print("-" * 30)
        if seat == 0:
            print('죄송합니다.')
            print(movie, '잔여석이 없습니다.')
            print("-" * 30)
    else:
        print(movie, '영화 정보를 찾을 수 없습니다.')
        print("-" * 30)
````
