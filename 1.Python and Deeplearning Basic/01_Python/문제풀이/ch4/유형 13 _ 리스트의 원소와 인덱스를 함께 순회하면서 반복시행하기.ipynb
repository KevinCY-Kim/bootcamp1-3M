### 문제 24
주어진 문자열에서 대문자를 소문자로, 소문자를 대문자로 변환하여 새로운 문자열을 생성하는 프로그램을 작성해보세요.

### 정답

````python
text = "cHanGe uPPer tO LowER"
result = []

for i, char in enumerate(text):
    if char.islower():
        result.append(char.upper())
    elif char.isupper():
        result.append(char.lower())
    else:
        result.append(char)

result = ''.join(result)
print(result)
````

### 문제 25
array = [9, 10, 11, 8]일 때, 주어진 리스트에서 가장 작은 값을 찾아 그 값과 해당 값의 인덱스를 출력하는 프로그램을 작성해보세요.

(단, for문을 이용하여 코드를 작성해주세요.)




### 정답

````python
array = [9, 10, 11, 8]

min_value = array[0]
min_index = 0

for i in range(1, len(array)):
    if array[i] < min_value:
        min_value = array[i]
        min_index = i

print(f"가장 작은 값: {min_value}, 인덱스: {min_index}")
````
