### 문제 12 (사고력 ***)
아래 코드를 읽고 출력 결과를 맞춰보세요

````python
s = "10 Z 20 Z 1"

answer = 0
s = s.split() # [10, Z, 20, Z, 1] # 5개짜리 리스트
res = []
for i in s: # 5바퀴
    # 10, Z, 20
    if i == 'Z':
        if res: # 리스트 원소 존재
            target = res.pop()
            answer -= target # 10 - 10 + 20 - 20
    else:
        i = int(i) # 20
        answer += i  # 20
        res.append(i) # res = [10]

answer
````

### 문제 24
주어진 2차원 리스트는 학생들의 국,영,수 점수를 나타냅니다. 점수에 따라 평가 등급을 매기고, 이를 결과 리스트에 저장해야합니다. 점수가 90점 이상이면서 평균 점수도 90점 이상인 경우 'A', 점수는 90점 이상이지만 평균 점수가 90점 미만인 경우 'B', 나머지 경우에는 'C' 일 때, 학생들의 시험 점수를 평가하는 프로그램을 작성해보세요.

1. 주어진 학생들의 점수들을 반복하면서, 각 학생의 최고 점수가 90점 이상인지 확인합니다.
2. 최고 점수가 90점 이상인 경우에만 평균 점수를 계산하여 평균이 90점 이상인지 확인합니다.
3. 조건에 따라 'A','B','C'등급을 부여하고 결과 리스트에 추가하여 출력합니다.

### 정답

````python
scores = [
  [90, 85, 80],
  [88, 92, 77],
  [100, 95, 90],
  [85, 90, 95],
  [75, 80, 85],
]

answers = []

for score in scores:
    if max(score) >= 90:
        if sum(score) / len(score) >= 90:
            answers.append("A")
        else:
            answers.append("B")
    else:
        answers.append("C")

print(answers)
````

### 문제 30
다음은 아이디와 패스워드가 포함된 리스트와 데이터베이스 리스트입니다.
다음 규칙에 맞도록 코드를 작성해주세요.
- 아이디와 패스워드가 일치하는 경우에는 '로그인 되었습니다.'를 출력
- 아이디는 있으나 패스워드가 동일하지 않은 경우 '패스워드가 틀렸습니다.'를 출력
- 아이디가 존재하지 않는 경우에는 '존재하지 않는 아이디입니다.'를 출력
```
id_pw = ["john", "abc123"]
db = [["hannah", "123"], ["jenny", "pw456"], ["james", "abc123"]]
```


````python
id_pw = ["john", "abc123"]
db = [["hannah", "123"], ["jenny", "pw456"], ["james", "abc123"]]

#아이디와 패스워드가 일치하는 경우에는 '로그인 되었습니다.'를 출력
#아이디는 있으나 패스워드가 동일하지 않은 경우 '패스워드가 틀렸습니다.'를 출력
#아이디가 존재하지 않는 경우에는 '존재하지 않는 아이디입니다.'를 출력

answer = "존재하지 않는 아이디입니다."

for d in db:
    if d[0] == id_pw[0]:
        if d[1] == id_pw[1]:
            answer = "로그인 되었습니다."
        else:
            answer = "패스워드가 틀렸습니다."
print(answer)
````

### 정답

````python
id_pw = ["john", "abc123"]
db = [["hannah", "123"], ["jenny", "pw456"], ["james", "abc123"]]

answer = "존재하지 않는 아이디입니다."
for i in db:
    # id 여부 확인
    if id_pw[0] == i[0]:
        if id_pw[1] == i[1]:
            answer = "로그인 되었습니다."
        else:
            answer = "패스워드가 틀렸습니다."

print(answer)
````
