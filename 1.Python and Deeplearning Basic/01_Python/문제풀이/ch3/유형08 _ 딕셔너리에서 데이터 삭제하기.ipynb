###문제 30




떡볶이집 알바생 유리는 사장님으로부터 로제떡볶이 메뉴가 사라짐을 듣고 메뉴판에서 '로제떡볶이'를 삭제합니다. 하지만 로제떡볶이를 너무 좋아했던 유리는 로제떡볶이의 정보들을 잘 기억해두고자 메모장에 맵기 단계와 조리시간이 15분임을 딕셔너리 형태로 작성해둡니다. 변경된 메뉴판과 유리 메모장에 다음과 같은 딕셔너리 형태로 작성되어 있게끔 코드를 작성하세요.
 - 주어진 딕셔너리에서 '로제떡볶이'를 pop() 메서드를 사용하여 삭제합니다.   
 - pop()을 통해 얻은 데이터에서 '가격' 키와 해당 값을 del 키워드를 사용하여 삭제합니다.  
 - 위 데이터에 조리시간 값을 추가하여 수정된 데이터를 출력합니다.  

```
<결과값>
메뉴판
menu = {
  '투움바떡볶이': {'맵기': [1, 2, 3], '가격': 7000},
  '일반떡볶이': {'맵기': [1, 2, 3, 4, 5], '가격': 5000}
}
유리 메모장
{'맵기': [1, 2, 3], '조리시간': '15분'}
```



````python
menu = {
  '로제떡볶이': {'맵기': [1, 2, 3], '가격': 7000},
  '투움바떡볶이': {'맵기': [1, 2, 3], '가격': 7000},
  '일반떡볶이': {'맵기': [1, 2, 3, 4, 5], '가격': 5000}
}
````

````python
#@title 정답
menu = {
  '로제떡볶이': {'맵기': [1, 2, 3], '가격': 7000},
  '투움바떡볶이': {'맵기': [1, 2, 3], '가격': 7000},
  '일반떡볶이': {'맵기': [1, 2, 3, 4, 5], '가격': 5000}
}

deleted_data = menu.pop('로제떡볶이')
del deleted_data['가격']

deleted_data['조리시간'] = '15분'
print(deleted_data)
print(menu)
````
