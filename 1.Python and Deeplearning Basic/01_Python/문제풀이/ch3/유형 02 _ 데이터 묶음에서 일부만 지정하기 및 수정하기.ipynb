###문제 30
다음은 학생별 국어,영어,수학점수를 표로 작성한 것이다.


|이름|국어점수|영어점수|수학점수|
|:----:|:--------:|:--------:|:--------:|
|Alice|80|90|85|
|Bob|70|60|80|
|Charlie|95|80|90|
|David|75|85|70|

 grades 리스트 정보를 활용해  **'Alice의 평균은 85.0입니다'**를 출력되도록 코드를 작성해보세요.  

````python
grades = [
    ["Alice", 80, 90, 85],
    ["Bob", 70, 60, 80],
    ["Charlie", 95, 80, 90],
    ["David", 75, 85, 70],
]
````

````python
#@title 정답
grades = [
    ["Alice", 80, 90, 85],
    ["Bob", 70, 60, 80],
    ["Charlie", 95, 80, 90],
    ["David", 75, 85, 70],
]

name_alice = grades[0][0]
alice_average = sum(grades[0][1:])/len(grades[0][1:])

print(name_alice,'의 평균은',alice_average,'입니다')
````
