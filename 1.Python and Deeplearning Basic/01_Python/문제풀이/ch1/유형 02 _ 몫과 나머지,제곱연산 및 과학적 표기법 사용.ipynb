### 문제 심화
자판기에서 음료수를 구매하고자 1000원짜리 음료수를 선택했고 5,000원권권을 투입했다. 1,000원권 4장으로 4,000원을 거슬러 줄 수 있는 코드를 아래를 참고하여 ( )을 완성해주세요.  

```
selected_price = 1000
input_money = 5000

change = (           ) - selected_price
thousands = change // (     )
remainder = change %  (     )

print("거스름돈:", change, "원")
print("1,000원권:", thousands, "장")
print("나머지:", remainder, "원")

```



````python
# @title 정답
selected_price = 1000
input_money = 5000

change = input_money - selected_price
thousands = change // 1000
remainder = change % 1000

print("거스름돈:", change, "원")
print("1,000원권:", thousands, "장")
print("나머지:", remainder, "원")
````

### 문제 심화
반지름이 2.5cm인 원의 넓이를 구하고자한다. 단, 원주율은 3.14로 계산합니다. 넓이는 과학적 표기법으로 출력된다. 아래의 코드를 보고 빈칸을 완성해주세요.

```
radius = 2.5 # 반지름(cm)
pi = 3.14 # 원주율

# 원의 넓이 계산
area = (    ) * (     ) **2

# 과학적 표기법으로 출력
result = print("원의 넓이(과학적 표기법):", format(area, ".4e"))

```




````python
# @title 정답
radius = 2.5 # 반지름(cm)
pi = 3.141592653589793 # 원주율

# 원의 넓이 계산
area = pi * radius ** 2

# 제곱형식으로 출력
print("원의 넓이(제곱형식):", area)

# 과학적 표기법으로 출력
print("원의 넓이(과학적 표기법):", format(area, ".4e"))
````
