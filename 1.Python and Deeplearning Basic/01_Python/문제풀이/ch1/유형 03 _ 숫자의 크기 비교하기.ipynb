### 문제 30

a,b,c = 10,27,3일 때,

1. num1은 a와 b를 더한 값에 c를 나눈 몫
2. num2는 a와 c를 곱한 값에 b를 나눈 나머지
3. num3는 b와 c를 뺀 값에 10을 나눈 값

result의 값은 num1이 num3보다 크거나 같고 num1 보다 num2가 크거나 num2는 num3가 다르다. 이를 코드로 작성해보세요.
