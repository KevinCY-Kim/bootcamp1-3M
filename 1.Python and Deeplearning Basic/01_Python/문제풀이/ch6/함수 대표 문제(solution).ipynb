### 문제 2 (4-22)
'yyyy-mm-dd' 형식의 날짜 문자열을 받아 'mm/dd/yyyy' 형식으로 변환하는 format_date 함수를 활용하여

'2023-06-30' 형식의 날짜를 '06/30/2023' 형식으로 변환하여 출력해보세요.

### 정답

````python
def format_date(date):
    year, month, day = date.split('-')
    return f"{month}/{day}/{year}"

date_str = '2023-06-30'
formatted_date = format_date(date_str)
print(formatted_date)
````

### 문제 3 (4-29)
( 해당 문제는 인공지능 자연어처리 분야에서 다루는 소재를 바탕으로 만들었습니다.)

word_list와 morpheme_dict을 입력 인자로 받아 주어진 단어 리스트를 품사 태그로 반환하는 translate_words 함수를 작성하세요.


- word_list: 분석할 단어들이 들어 있는 리스트입니다.
- tag_map: 각 단어와 그에 해당하는 품사를 연결하는 사전입니다.
    - word_list의 각 단어를 순회하면서, 해당 단어가 tag_map 사전에 있으면 그 품사를, 없으면 'UNK'을 출력합니다.

```
# 예시 입력

word_list = ["she", "enjoy", "summer", "holiday", "abroad"]
tag_map = {"she": "PRON", "enjoy": "VB", "summer": "NN", "holiday": "NN"}

tags = morphemes(sentence, tag_map)
print(tags)

# 예상 출력

['PRON', 'VB', 'NN', 'NN', 'UNK']
```

### 정답

````python
def morphemes(word_list, morpheme_dict):
    tags = []
    for word in word_list:
        if word in morpheme_dict:
            tags.append(morpheme_dict[word])
        else:
            tags.append("UNK")  # 알 수 없는 단어 처리
    return tags

sentence = ["she", "enjoy", "summer", "holiday", "abroad"]

# 단어를 형태소 태그로 매핑하는 사전
tag_map = {
    "she": "PRON",   # PRON: 대명사, 사람이나 사물을 대체하여 사용
    "enjoy": "VB",   # VB: 동사, 행동이나 상태를 나타냄
    "summer": "NN",  # NN: 명사, 사람, 장소, 사물 또는 개념을 나타냄
    "holiday": "NN", # NN: 명사, 사람, 장소, 사물 또는 개념을 나타냄
}

# 단어 리스트를 형태소 태그로 변환
tags = morphemes(sentence, tag_map)
print(tags)
````

### 문제 4 (5-22)
process_order 함수는 item, quantity, payment_method 세 개의 입력 인자를 받아 주문을 처리합니다.

- item은 주문하는 상품
- quantity는 주문 수량
- payment_method는 결제 방법 (default value : '신용카드')

아래와 같은 결과가 출력되도록 코드를 작성해보세요.



```
커피 2개 주문이 완료되었습니다.
결제 방법: 신용카드
도서 1개 주문이 완료되었습니다.
결제 방법: 계좌이체
```



### 정답

````python
def process_order(item, quantity, payment_method="신용카드"):
    print(f"{item} {quantity}개 주문이 완료되었습니다.")
    print(f"결제 방법: {payment_method}")

process_order("커피", 2)
process_order("도서", 1, "계좌이체")
````

### 문제 5 (6-27)
make_sound 함수는 동물의 소리를 출력하는 함수로 animal이라는 입력 인자를 받으며, 기본값으로 "개"가 설정되어 있습니다. 주어진 animal 값에 따라 해당 동물의 소리를 출력해보세요.

-  "개"일 경우 "멍멍!", "고양이"일 경우 "야옹!", 그 외의 값일 경우 "알 수 없는 동물입니다."라는 메시지를 출력합니다.

### 정답

````python
def make_sound(animal="개"):
    if animal == "개":
        print("멍멍!")
    elif animal == "고양이":
        print("야옹!")
    else:
        print("알 수 없는 동물입니다.")

make_sound()
make_sound("고양이")
make_sound("사자")
````

### 문제 6 (7-25)
주어진 숫자(number)가 양수인지, 음수인지, 아니면 0인지 확인할 수 있도록 코드를 람다 함수를 사용하여 구현하세요.
- 양수인 경우 "양수입니다."를 출력하세요.
- 음수인 경우 "음수입니다."를 출력하세요.
- 0인 경우 "0입니다."를 출력하세요.


### 정답

````python
check_number = lambda number: "양수입니다." if number > 0 else ("음수입니다." if number < 0 else "0입니다.")

result = check_number(0)
result
````

### 문제 7 (7-29)
주어진 파일명 리스트 filenames와 확장자 extension가 주어졌을 때, 해당 확장자를 가진 파일명만 필터링하여 출력하는 파이썬 스크립트를 작성하세요.

이 작업을 수행하기 위해 람다 함수를 사용하고, 리스트 컴프리헨션을 이용하세요.

- extension: 필터링할 파일의 확장자 (예: 'jpg')
```
filenames = ['readme.txt', 'car.jpg', 'airplane.jpg', 'person.png', 'flower.jpg', 'bicycle.bmp', 'cat.jpg']
```



### 정답

````python
print_files_with_extension = lambda filenames, extension: print([filename for filename in filenames if filename.endswith(extension)])

filenames = ['readme.txt', 'car.jpg', 'airplane.jpg', 'person.png', 'flower.jpg', 'bicycle.bmp', 'cat.jpg']
extension = 'jpg'

print_files_with_extension(filenames, extension)
````
