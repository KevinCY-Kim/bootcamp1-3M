## 기사제목

```python
# 1. get 생략
# 2. select
soup = BeautifulSoup(resp.content, 'lxml')
soup
```

#### 타겟태그 자체가 클래스, 아이디 속성이 없을 때
- 계층구조로 독특한 특성 찾기 : 상위태그 중 클래스, 아이디 속성을 가진 상위태그를 표현함으로써
- 상위태그와 하위태그를 구분하는 구분자 : 공백

#### 클래스에 공백이 포함되어 있을 때
- 클래스 값이 여러개라는 의미
- select 구문에서 작성 시 : 모든 클래스 값 앞에 .을 붙인다.
- 필수로 모두 다 적어야 하는 것은 아니다. 대표적으로 하나만 명시해도 된다. 단, 그럴 경우, 우리가 원하는 태그만 가져오지는 못할 수 있다. (만약, 원하는 태그집단이 여러 클래스 값을 가지고 있는데, 하나만 일치하고 다른 값이 다르면 일치하는 클래스 값만 표현하면 된다.)

```python
# 2-1) 태그 파싱
news_tags = soup.select('strong.tit-g.clamp-g a')
len(news_tags)
```

```python
# 2-2) 데이터 파싱
for news_tag in news_tags:
    print(news_tag.text)
```

### href 속성값 가져오기

```python
for news_tag in news_tags:
    print(news_tag['href']) # 딕셔너리 키값 명시하듯이
```

### 딕셔너리에 기사 제목과 url 을 저장해보기

```python
news_info = {}

titles = []
urls = []

for news_tag in news_tags:
    titles.append(news_tag.text)
    urls.append(news_tag['href'])

news_info['title'] = titles
news_info['url'] = urls
news_info
```

```python
news_info = {'title' : [], 'url' : []}

for news_tag in news_tags:
    news_info['title'].append(news_tag.text)
    news_info['url'].append(news_tag['href'])

news_info
```

### 크롤링 정보 CSV 파일로 저장하기

```python
import pandas as pd
df_news = pd.DataFrame(news_info)
df_news # pandas 에서 다루는 dataframe 타입
```

```python
type(df_news)
```

```python
df_news.to_csv('삼성전자뉴스정보.csv', encoding='utf-8-sig', index=False)
```

## 10페이지에 걸쳐서 뉴스기사 정보 100개 가져오기

```python
from tqdm.notebook import tqdm
import time
```

```python
news_100 = {'title' : [], 'url' : []}

for i in tqdm(range(10), desc='뉴스정보'):
    url = f'https://search.daum.net/search?nil_suggest=btn&w=news&DA=SBC&cluster=y&q=%EC%82%AC%EA%B3%BC={i+1}'
    resp = requests.get(url)
    time.sleep(1)
    soup = BeautifulSoup(resp.content, 'lxml')
    # 2-1) 태그 파싱
    news_tags = soup.select('strong.tit-g.clamp-g a')
    # 2-2) 데이터 파싱
    for news_tag in news_tags:
        news_100['title'].append(news_tag.text)
        news_100['url'].append(news_tag['href'])
news_100
```
