
### 2-2) 아래 사항 선택 가능하도록 함수화하기
- 원하는 웹툰
- 최신 에피소드 / 전체 에피소드 여부
- 베스트 댓글 / 전체 댓글 여부

```
def wbtn_crawling(title, current_only=True, best_only=True): # False : total
    contents = []

    title_id = webtoon_db[title]['title_id']
    current_no = webtoon_db[title]['current_no']

    # 최신화만 : start = 최신화 no / 전체화 : start = 1
    start = current_no if current_only else 1
    # 베스트 댓글 / 전체 댓글 여부
    sort = 'best' if best_only else 'new'

    for no in tqdm(range(start, current_no+1), desc='episode'):
        url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json'
        params['objectId'] = f'{title_id}_{current_no}'
        params['sort'] = sort # best or new

        resp = requests.get(url, params=params, headers=headers)

        jsn = json.loads(resp.text[10:-2])
        # best : 항상 1 / new : N
        total_pages = jsn['result']['pageModel']['totalPages']

        for page in tqdm(range(1, total_pages+1), desc=f'({no}-epi)review_page'):
            url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json'

            params['page'] = page

            resp = requests.get(url, params=params, headers=headers)

            dic_resp = json.loads(resp.text[10:-2])
            commentList = dic_resp['result']['commentList']
            for comment in commentList:
                contents.append(comment['contents'])
    return contents
```

```
# 원하는 입력 파라미터 설정할 때 정상작동하는지 확인
title = input('원하는 웹툰 입력 : ')
current_only = bool(int(input('최신화(1)/전체화(0) 여부 선택 : ')))
best_only = bool(int(input('베스트댓글(1)/전체댓글(0) 여부 선택 : ')))

title, current_only, best_only
```

```
reviews = wbtn_crawling(title=title,
                        current_only=current_only,
                        best_only=best_only)
reviews
```

### 3-1) 아래 사항을 선택 가능하도록 함수화하기
- 최신 에피소드 / 전체 에피소드 여부
- 베스트 댓글 / 전체 댓글 여부

```
def total_wbtn_crawler(current_only=True, best_only=True)
    for webtoon_info in webtoon_db.values():
        title_id, current_no = webtoon_info['title_id'], webtoon_info['current_no']

        ### 최신 에피소드 / 전체 에피소드 여부
        start = current_no if current_only else 1

        for no in range(start, current_no+1):
            # total pages 가져오기
            # best : 1 / new : N
            url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json'
            params['objectId'] = f'{title_id}_{no}'

            ### 베스트 댓글 / 전체 댓글 여부
            params['sort'] = 'best' if best_only else 'new'

            resp = requests.get(url, headers=headers)
            jsn = json.loads(resp.text[10:-2])
            total_pages = jsn['result']['pageModel']['totalPages']

            for page in tqdm(range(1, total_pages+1), desc=f'({no}-epi)review_page'):
                url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json'

                params['page'] = page

                resp = requests.get(url, params=params, headers=headers)

                print(resp.text)
                # dict 변환하여 파싱부분 위와 동일하므로 생략
```
