### 2-4. 종합 카테고리 모든 도서의 제목, 작가, 가격 정보 csv로 저장하기
- 가격은 숫자로 변환하여 저장하자.

```python
from tqdm.notebook import tqdm
```

```python
tits, prcs, auts = [], [], []

for i in tqdm(range(20)):
    page = i+1
    url = 'https://www.aladin.co.kr/shop/common/wbest.aspx'

    params = {
        'BestType':'Bestseller',
        'BranchType':1,
        'CID':0,
        'page':page,
        'cnt':1000,
        'SortOrder':1
        }
    resp = requests.get(url, params=params)
    print(resp.url)

    # 제목, 가격
    soup = BeautifulSoup(resp.content, 'lxml')
    a_tags = soup.select('a.bo3')
    # len(a_tags)

    for a_tag in a_tags:
        tits.append(a_tag.text)
    # tits

    em_tags = soup.select('span.ss_p2 em')
    # len(em_tags)

    for em_tag in em_tags:
        price = em_tag.text # '18,000원' -> 18000
        price = int(price.replace('원', '').replace(',', ''))
        prcs.append(price)
    # prcs

    # 작가
    bbox_tags = soup.select('div.ss_book_box')

    blst_tags = []
    for bbox_tag in bbox_tags:
        blst_tag = bbox_tag.select('div.ss_book_list')[0]
        blst_tags.append(blst_tag)
    # len(blst_tags)

    for idx, blst_tag in enumerate(blst_tags):
        li_tags = blst_tag.select('li')

        li1_span = li_tags[0].select('span')[0]
        # print(li1_span['class'])
        if li1_span['class'][0] == 'tit_category':
            author = li_tags[1].text.split('|')[0]
        else:
            author = li_tags[2].text.split('|')[0]
        auts.append(author)
    # auts
```

```python
dic_aladin = {'title' : tits, 'author': auts, 'price': prcs}
dic_aladin
```

```python
import pandas as pd

df_aladin = pd.DataFrame(dic_aladin)
df_aladin.to_csv('알라딘_종합도서.csv',
                 encoding='utf-8-sig',
                 index=False)
```

## 4. 알라딘 도서 섬네일 가져오기

#### 이미지 크롤링의 이해  
- 크롤링 단계 리마인드
    1. 페이지 가져오기
    2. 태그 속 정보 파싱하기   
        2.1 태그 파싱 (html에서)  
        2.2 정보 파싱 (태그에서)

- 이미지 정보의 특징
    - 이미지 태그 속에 이미지는 없다. (대신 img_src가 있다.)


- 이미지 크롤링 전략
    - 태그 파싱까지는 동일
    - 이미지 파싱은 다음과 같이 진행
        1. 1단계 : 섬네일 이미지 태그에서 img src 가져오기
        2. 2단계 : 이미지 src에 requests get하여 img 가져오기

#### 4-3) 모든 카테고리의 도서 섬네일 이미지 저장하기

```python
import requests
from bs4 import BeautifulSoup
import os
from tqdm.notebook import tqdm
```

##### 카테고리 수집

```python
cids, cats = [], []
```

```python
'/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=55890'.split('CID=')[-1]
```

```python
url = 'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0'
resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'lxml')
a_tags = soup.select('ul.best_left_ul li a')
print(len(a_tags))
for a_tag in a_tags:
    cats.append(a_tag.text.replace('/', ''))
    cids.append(a_tag['href'].split('CID=')[-1])
cats, cids
```

```python
dir = 'aladin_top50'
os.mkdir(dir)
```

```python
for cat in cats:
    os.mkdir(dir+'/'+cat)
```

```python
for cid, cat in zip(tqdm(cids), cats):
    url = f'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID={cid}'
    resp = requests.get(url)

    soup = BeautifulSoup(resp.content, 'lxml')
    img_tags = soup.select('div.flipcover_in img')

    # print(len(img_tags))
    for idx, img_tag in enumerate(tqdm(img_tags, desc=cat)):
        # src parsing
        # get(src)
        resp = requests.get(img_tag['src'])

        # resp.content 를 해당 카테고리 폴더에 저장
        with open(f'aladin_top50/{cat}/{cat}_{idx:02}.jpg', 'wb') as f:
            f.write(resp.content)
```
