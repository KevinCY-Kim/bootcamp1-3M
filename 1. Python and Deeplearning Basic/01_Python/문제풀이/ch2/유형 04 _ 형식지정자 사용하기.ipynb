###문제 24
주어진 문자열을 활용하여 '박가을고객님의 나이는 25이며 성별은 여성입니다.'를 출력해보세요. (단, format() 사용)

- format(age,name,gender) 순으로 작성

````python
age = 25
name = "박가을"
gender = '여성'

````

````python
# @title 정답
age = 25
name = "박가을"
gender = '여성'

result = "{1}고객님의 나이는 {0}이며 성별은 {2}입니다.".format(age, name,gender)
result
````

###문제 25
주어진 문자열을 활용하여 아래와 같이 출력해보세요. (단, format() 사용)



```
car.jpg의 카테고리는 car이고 두 번째 파일명은 car_02이 되고 확장자는 jpg입니다.
```



````python
filename = 'car.jpg'
category = filename[:3]
extension = filename[-3:]
num = 2
````

````python
# @title 정답
filename = 'car.jpg'
category = filename[:3]
extension = filename[-3:]
num = 2

result = '{}의 카테고리는 {}이고 두 번째 파일명은 car_{:02d}이 되고 확장자는 {}입니다.'.format(filename, category, num, extension)
print(result)
````

###문제 30
주어진 문자열과 split(),인덱싱,f-string을 사용하여 아래와 같이 출력해보세요.



```
이미지의 가로너비는 800 이며, 세로 높이는 600입니다
```



````python
image = '800,600'

````

````python
# @title 정답
image = '800,600'
width, height = image.split(',')[0], image.split(',')[1]

result = f'이미지의 가로너비는 {width} 이며, 세로 높이는 {height}입니다'
result
````
