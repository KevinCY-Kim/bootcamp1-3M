### 문제 심화
: 물건의 가격(price)이 15,000원이고, 소비세(tax)율이 10%일 때,
1. 'tax_amount' 변수에는 세금액이 저장되어야 합니다.
2. 'total_price' 변수에는 세포함 총 가격이 저장되어야 합니다.

아래의 코드를 참고하여 'tax_amount'와 'total_price'를 계산하는 코드를 작성해보세요.



```
# price = 15000
tax = 0.1

tax_amount = ?
total_price = ?

```


````python
# @title 정답

price = 15000
tax = 0.1

tax_amount = price * tax
total_price = price + tax_amount

print("세금액:", tax_amount)
print("세포함 총 가격:", total_price)
````
