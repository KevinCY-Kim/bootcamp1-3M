### 문제 3 (3-21)

주어진 리뷰 리스트에서 "너무"라는 단어는 "완전"이라는 단어로 변경하고, '별로'라는 단어가 들어간 리뷰는 삭제하고자 한다. 유리를 위해 파이썬 코드를 작성해주세요.


````python
reviews = ["음식이 너무 짜요", "매장이 너무 시끄러워요", "가격이 너무 비쌉니다", "서비스가 너무 좋아요!",'서비스가 별로에요']
````

````python
#@title 정답
reviews = ["음식이 너무 짜요", "매장이 너무 시끄러워요", "가격이 너무 비쌉니다", "서비스가 너무 좋아요!",'서비스가 별로에요']

reviews = [review.replace("너무", "완전") if "너무" in review else review for review in reviews if "별로" not in review]

print(reviews)
````

### 문제 5 (5-24)
 주어진 중첩된 리스트를 평탄화하여 하나의 단일 리스트로 변환하는 코드를 작성해보세요.

````python
nested_list = [['apple', 'banana'], ['cherry', 'date'], ['elderberry', 'fig']]
````

````python
#@title 정답
nested_list = [['apple', 'banana'], ['cherry', 'date'], ['elderberry', 'fig']]
flattened_list = [fruit for sublist in nested_list for fruit in sublist]
print(flattened_list)
````

### 문제 6 (6-23)
 주어진 리스트를 활용하여 요일에 따른 작업 일정 정보를 담은 딕셔너리를 생성해보세요.

````python
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
tasks = ["Meeting", "Presentation", "Report", "Training", "Planning"]
````

````python
# 요일과 작업 일정
#@title 정답
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
tasks = ["Meeting", "Presentation", "Report", "Training", "Planning"]
schedule_dict = {day: task for day, task in zip(weekdays, tasks)}
print(schedule_dict)
````
