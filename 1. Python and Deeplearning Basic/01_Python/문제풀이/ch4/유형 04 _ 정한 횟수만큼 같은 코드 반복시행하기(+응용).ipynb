### 문제 30
ticket_prices = {'성인': 25000, '소아': 20000, '유아': 15000}일 때, "소아 가격: 20000원"이 출력되도록 코드를 작성해보세요.



````python
ticket_prices = {'성인': 25000, '소아': 20000, '유아': 15000}

for category, price in ticket_prices.items():
    if category == '소아':
        print(f"{category} 가격: {price}원")
````

### 정답

````python
ticket_prices = {'성인': 25000, '소아': 20000, '유아': 15000}

for i in range(len(ticket_prices)):
  category = list(ticket_prices.keys())[i]
  if category == '소아':
    price = ticket_prices[category]
    print(f"{category} 가격: {price}원")
````
