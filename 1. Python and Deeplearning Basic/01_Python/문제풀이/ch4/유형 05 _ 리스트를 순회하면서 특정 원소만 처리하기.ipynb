### 문제 30
아래 data 리스트에서 삼성전자 주가 평균과 LG전자 주가 평균을 구하여 작성해보세요.

```python
# '회사명', '지역', '주식 보유 기간', '주당 가격'
data = [
('삼성전자', '서울', 20, 10000),
('LG전자', '서울', 25, 12000),
('SK하이닉스', '서울', 30, 15000),
('삼성전자', '경기', 30, 13000),
('LG전자', '경기', 35, 16000),
('SK하이닉스', '경기', 40, 18000),
]

samsung = []
lg = []

for d in data:
    if d[0] == '삼성전자':
        samsung.append(d[3])
    if d[0] == 'LG전자':
        lg.append(d[3])

samsung_avg = sum(samsung) / len(samsung)
lg_avg = sum(lg) / len(lg)

print("삼성전자 주가 평균:", samsung_avg)
print("LG전자 주가 평균:", lg_avg)
```
