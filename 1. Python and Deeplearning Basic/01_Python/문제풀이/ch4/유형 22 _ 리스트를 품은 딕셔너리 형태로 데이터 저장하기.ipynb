### 문제 22
다음과 같이 환율(exchange_rates)과 금액 정보(amounts)가 주어졌을 때,

금액 정보(amounts) 금액을 환전한 결과를 딕셔너리 형태로 저장하는 코드를 작성해보세요.
- 환전한 결과는 소숫점 첫째 자리까지 출력하세요.
- 단, 주어진 환율에 해당하는 통화가 없는 경우 딕셔너리에 키를 추가하고 값으로 None을 할당하세요.

```
# 출력 결과

{'USD': [1500000, 1348.5], 'EUR': [1250000, 922.5], 'JPY': [620000, 61249.8], 'CNY': [800000, None]}
```

````python
exchange_rates = {
    "USD": 0.000899,
    "EUR": 0.000738,
    "JPY": 0.09879,
}
amounts = [(1500000, "USD"), (1250000, "EUR"), (620000, "JPY"), (800000, "CNY")]

converted_amounts = {}
#환전한 결과는 소숫점 첫째 자리까지 출력하세요.
#단, 주어진 환율에 해당하는 통화가 없는 경우 딕셔너리에 키를 추가하고 값으로 None을 할당하세요.
for amount, currency in amounts:
    exchange_rate = exchange_rates.get(currency)
    if exchange_rate is not None:
        converted_amount =round(amount * exchange_rate, 1)
        converted_amounts[currency] = [amount, converted_amount]
    else:
        converted_amounts[currency] = [amount, None]
print(converted_amounts)
````

### 정답

### 문제 25
주어진 문자열 리스트에서 'e'를 포함하는 문자열과 포함하지 않는 문자열을 딕셔너리로 분류하는 코드를 작성해보세요.
- 딕셔너리: 'e'를 포함하는 문자열을 키 'has_e'에 저장하고, 포함하지 않는 문자열을 키 'no_e'에 저장하세요.
```
num = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
```
출력 결과
```
{'no_e': ['two', 'four', 'six'], 'has_e': ['zero', 'one', 'three', 'five', 'seven', 'eight', 'nine']}
```



````python
num = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']

no_e = []
has_e = []

for n in num:
    found_e = False
    for char in n:
        if char == 'e':
            found_e = True
            break
    if found_e:
        has_e.append(n)
    else:
        no_e.append(n)
result = {'no_e' : no_e, 'has_e' : has_e}
print(result)
````

### 정답

````python
num = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']

no_e = []
has_e = []

for n in num:
    found_e = False
    for char in n:
        if char == 'e':
            found_e = True
            break
    if found_e:
        has_e.append(n)
    else:
        no_e.append(n)

result = {'no_e': no_e, 'has_e': has_e}

print(result)
````

### 문제 28
다음과 같은 점수 데이터가 주어졌을 때, 총점을 기준으로 내림차순으로 데이터를 정렬하는 코드를 작성해보세요.
```
score_data = [{'name': 'Alice', 'Math': 80, 'Science': 75, 'English': 90},
              {'name': 'Bob', 'Math': 90, 'Science': 85, 'English': 80},
              {'name': 'Charlie', 'Math': 85, 'Science': 90, 'English': 85}]
```
            
```
# 출력 결과  

[{'name': 'Charlie', 'MSE_score': [85, 90, 85], 'total_score': 260}, {'name': 'Bob', 'MSE_score': [90, 85, 80], 'total_score': 255}, {'name': 'Alice', 'MSE_score': [80, 75, 90], 'total_score': 245}]
```




### 정답

````python
# 삽입 정렬
score_data = [{'name': 'Alice', 'Math': 80, 'Science': 75, 'English': 90},
              {'name': 'Bob', 'Math': 90, 'Science': 85, 'English': 80},
              {'name': 'Charlie', 'Math': 85, 'Science': 90, 'English': 85}]

score_dict = []
for d in score_data:
    name = d['name']
    scores = [d['Math'], d['Science'], d['English']]
    total_score = sum(scores)
    score_dict.append({'name': name, 'MSE_score': scores, 'total_score': total_score})

for i in range(len(score_dict)):
    max_idx = i
    for j in range(i+1, len(score_dict)):
        if score_dict[j]['total_score'] > score_dict[max_idx]['total_score']:
            max_idx = j
    score_dict[i], score_dict[max_idx] = score_dict[max_idx], score_dict[i]

print(score_dict)
````
