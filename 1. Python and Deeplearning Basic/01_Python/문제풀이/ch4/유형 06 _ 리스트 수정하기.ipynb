### 문제 23
lst = ["a", "b", "o", "b", "g", "a", "p"]
일 때, ['a', 'b', 'o', 'g', 'p']가 출력되도록 코드를 작성해보세요.

### 정답

````python
lst = ["a", "b", "o", "b", "g", "a", "p"]

for i in range(len(lst)-1, -1, -1):
    if lst[i] in lst[:i]:
        del lst[i]

print(lst)
````

### 문제 30

city_list에 있는 각 항목에 대해 반복문을 수행하여 **(도시 이름, 인구수(만 단위))** 형태의 튜플을 출력하고자 합니다.

이때, 인구 수는 '만' 단위 값을 얻기 위해 네 자리 숫자를 제외한 값을 출력하는 코드를 작성하세요.
```
cities_populations = {
    'Seoul': 9720846,
    'Tokyo': 13929286,
    'New York': 8336817,
    'Mumbai': 12442373,
    'Moscow': 12506468,
    'Sao Paulo': 12252023
}
```
#### 출력 결과
```
('Seoul', 972)
('Tokyo', 1392)
('New York', 833)
('Mumbai', 1244)
('Moscow', 1250)
('Sao Paulo', 1225)
```


### 정답

````python
cities_populations = {
    'Seoul': 9720846,
    'Tokyo': 13929286,
    'New York': 8336817,
    'Mumbai': 12442373,
    'Moscow': 12506468,
    'Sao Paulo': 12252023
}

for city, pop in cities_populations.items():
    print((city, pop // 10000))
````
