````python
string1 = ['인', '어', '공', '주']
string2 = ['공', '기', '층']
string3 = ['인', '지']
string4 = ['재', '능']

result, count = [], 0
found_innovation = False

total_combinations = len(string1) * len(string2) * len(string3) * len(string4)

while count < total_combinations and not found_innovation:
    i = count // (len(string2) * len(string3) * len(string4))
    j = (count // (len(string3) * len(string4))) % len(string2)
    k = (count // len(string4)) % len(string3)
    z = count % len(string4)

    combination = string1[i] + string2[j] + string3[k] + string4[z]
    result.append(combination)
    count += 1

    if combination == '인공지능':
        print(f'{count} : {result}')
        found_innovation = True
        break

if not found_innovation:
    print("인공지능을 찾을 수 없습니다.")
````

### 문제 19
아래 코드를 읽고 출력 결과를 맞춰보세요

````python
text = '/content/image/'
categories = ['USA', 'JAPAN', 'ITALY', 'KOREA', 'HONGKONG']
files = ['01.jpg', '02.jpg', '03.jpg']

result, count = [], 0
batch_size = 3  # 출력을 끊을 개수

for c in categories:
    for f in files:
        count += 1
        result.append(text + c + '_' + f)

        if count % batch_size == 0:
            if count % len(files) == 0:
                batch_end = result[-batch_size:]
            else:
                batch_end = result[-(count % len(files)):]
            print(f'{count} : {batch_end}')
            print("----- Batch End -----")

    if count % len(files) != 0:
        batch_end = result[-(count % len(files)):]
        print(f'{count} : {batch_end}')
        print("----- Batch End -----")

    print("----- Category End -----")
````

### 문제 27
두 개의 리스트(list1,list2)에서 각각 숫자를 선택하여 곱셈을 수행하고, 그 결과를 result 리스트에 추가하는 코드를 작성해보세요.

- result 리스트에 추가되는 원소는 순서대로 출력되지만 만약 곱셈 결과가 50이상이 되는 경우 반복문을 중단합니다.



### 정답

````python
list1 = [2, 4, 6, 8, 10]
list2 = [3, 6, 9, 12]

result, count = [], 0

for num1 in list1:
    for num2 in list2:
        count += 1
        result.append(num1 * num2)
        print(f"{count}: {result[-1]}")

        if result[-1] >= 50:
            break

    if result[-1] >= 50:
        break
````

### 문제 29
세 개의 문자열 리스트(string1,string2,string3)에서 각각 한개의 원소를 선택하여 합친 결과가 target_string과 일치하는 경우를 찾는 코드를 작성해보세요.

- 무한 루프를 돌면서 실행됩니다.
- 각 반복마다 count를 1씩 증가시키고, string1,string,string3에서 현재 인덱스에 해당하는 원소를 선택하여 합친 결과룰 result리스트에 추가합니다.
- 추가된 값이 target_string과 일치하는 경우에는 해당 결과를 출력합니다.
- k를 1 증가시키고, string3의 끝에 도달한 경우 k를 0으로 초기화하고, j를 1 증가시킵니다.
- j가 string2의 끝에 도달한 경우 j를 초기화하고 i를 1 증가시킵니다.
-i가 string1의 끝에 도달한 경우 무한 루프를 종료합니다.

### 정답

````python
string1 = ['a', 'b', 'c', 'd', 'e']
string2 = ['c', 't', 'a']
string3 = ['o', 's', 't']

result, count = [], 0
i, j, k = 0, 0, 0
target_string = 'cat'

while True:
    count += 1
    result.append(string1[i] + string2[j] + string3[k])

    if result[-1] == target_string:
        print(f'{count}: {result}')

    k += 1
    if k == len(string3):
        k = 0
        j += 1
        if j == len(string2):
            j = 0
            i += 1
            if i == len(string1):
                break
````

### 문제 30
파일 경로와 카테고리, 파일 리스트를 조합하여 batch를 형성하는 코드를 작성해보세요.

- 이중 반복문을 사용하여 categories와 files의 각 조합에 대해 실행됩니다.
- 각 조합으로 생성된 파일 경로를 result 리스트에 추가합니다.
- count를 1증가시키고, batch_size로 나눈 나머지가 0인 경우 배치의 끝에 도달한 것으로 간주합니다.
- 배치의 끝에 도달한 경우, result 리스트에서 해당 배치를 잘라내어 batch_end 변수에 저장하고 출력합니다.


### 정답

````python
text = '/content/image/'
categories = ['candle', 'basket', 'cushion']
files = ['01.jpg', '02.jpg', '03.jpg']

result, count = [], 0
batch_size = 3

for c in categories:
    for f in files:
        count += 1
        result.append(text + c + '_' + f)

        if count % batch_size == 0:
            if count % len(files) == 0:
                batch_end = result[-batch_size:]
            else:
                batch_end = result[-(count % len(files)):]
            print(f'{count} : {batch_end}')
            print("----- Batch End -----")

    if count % len(files) != 0:
        batch_end = result[-(count % len(files)):]
        print(f'{count} : {batch_end}')
        print("----- Batch End -----")

    print("----- Category End -----")
````
