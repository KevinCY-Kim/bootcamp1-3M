### 문제 19
아래 코드를 읽고 출력 결과를 맞춰보세요

````python
flights = {
    'BA213': {'출발 도시': '런던', '도착 도시': '뉴욕', '예정된 출발 시간': '15:30', '실제 출발 시간': '15:40', '상태': '온타임'},
    'AA101': {'출발 도시': '뉴욕', '도착 도시': '런던', '예정된 출발 시간': '17:00', '실제 출발 시간': '18:30', '상태': '지연'},
    'LH456': {'출발 도시': '중국', '도착 도시': '서울', '예정된 출발 시간': '20:00', '실제 출발 시간': '20:15', '상태': '온타임'},
    'EK301': {'출발 도시': '서울', '도착 도시': '일본', '예정된 출발 시간': '22:00', '실제 출발 시간': '22:00', '상태': '온타임'}
}

while 1:
    for flight_id, details in flights.items():
        scheduled_time = details['예정된 출발 시간'].split(':')
        actual_time = details['실제 출발 시간'].split(':')

        scheduled_hours = int(scheduled_time[0])
        scheduled_minutes = int(scheduled_time[1])
        actual_hours = int(actual_time[0])
        actual_minutes = int(actual_time[1])

        scheduled_total_minutes = scheduled_hours * 60 + scheduled_minutes
        actual_total_minutes = actual_hours * 60 + actual_minutes

        if actual_total_minutes - scheduled_total_minutes >= 60:
            print(f"항공편 {flight_id}는 지연되었습니다.")
            break
    else:
        print("모든 항공편이 정상적으로 운행 중입니다.")
        break
    break
````

### 문제 20

````python
# model의 키(key)는 머신러닝 모델의 이름을 나타내고, 값(value)은 각 모델의 성능 지표를 담고 있는 또 다른 딕셔너리입니다.
# 성능 지표 딕셔너리는 정확도(Accuracy), F1 스코어(F1 Score), 재현율(Recall), 정밀도(Precision)와 같은 중요한 평가 메트릭스를 포함합니다.

model = {
    'Decision Tree': {'Accuracy': 0.85, 'F1 Score': 0.82, 'Recall': 0.80, 'Precision': 0.85},
    'Random Forest': {'Accuracy': 0.90, 'F1 Score': 0.89, 'Recall': 0.87, 'Precision': 0.92},
    'SVM': {'Accuracy': 0.88, 'F1 Score': 0.86, 'Recall': 0.85, 'Precision': 0.88},
    'Neural Network': {'Accuracy': 0.92, 'F1 Score': 0.90, 'Recall': 0.91, 'Precision': 0.89}
}
target_model = 'Random Forest'

while 1:
    if target_model in model:
        metrics = model[target_model]
        print(f"모델 '{target_model}'의 성능 지표")
        for metric, value in metrics.items():
            print(f"- {metric}: {value}")
        break
    else:
        print(f"요청하신 모델 '{target_model}'의 성능 지표는 데이터베이스에 없습니다.")
        break
````

### 문제 24
머신러닝 모델을 훈련시키고 나면, 모델에 의해 계산된 각 **특성 중요도(Feature Importances)**를 검토하는 것이 중요합니다.

특성 중요도는 모델의 예측에 각 변수가 얼마나 영향을 미치는지를 나타냅니다.

주어진 특성 중요도 리스트에서 중요도가 너무 낮은 변수가 있을 경우,

해당 변수의 이름과 중요도를 출력하고 검사를 중단하는 프로그램을 작성하십시오.

- 무한 루프를 사용하여 features 리스트를 순회합니다.
- 각 변수의 중요도를 확인하고, 임계값 0.1 미만인 경우 그 변수의 이름과 중요도를 출력합니다.
- 임계값 미만인 중요도를 찾으면 반복문을 즉시 중단합니다.
- 모든 변수의 중요도가 임계값 이상인 경우, "모든 피처의 중요도가 적절합니다."라고 출력합니다.

```
features = ['age', 'income', 'education', 'marital_status', 'occupation']
importances = [0.2, 0.5, 0.05, 0.15, 0.1]  # 각 피처의 중요도
threshold = 0.1  # 중요도의 임계값
```

### 정답

````python
features = ['age', 'income', 'education', 'marital_status', 'occupation']
importances = [0.2, 0.5, 0.05, 0.15, 0.1]
threshold = 0.1

index = 0

while True:
    if importances[index] < threshold:
        print(f"피처 {features[index]}의 중요도가 낮습니다. (중요도: {importances[index]})")
        break
    index += 1

    if index >= len(features):
        print("모든 피처의 중요도가 적절합니다.")
        break
````

- 해당 정보를 출력한 후, 프로그램은 종료되어야 합니다.
```
login_attempts = {
    'user1': {'attempts': 3, 'times': ['12:01', '12:45', '13:00']},

    'user5': {'attempts': 1, 'times': ['11:00']}

```

출력 예시
```

사용자 user3의 로그인 시도가 임계값을 초과했습니다.

로그인 시간: ['22:15', '22:17', '22:20', '22:30', '22:35', '22:40', '23:00', '23:05']

```

### 정답

````python
login_attempts = {
    'user1': {'attempts': 3, 'times': ['12:01', '12:45', '13:00']},
    'user2': {'attempts': 5, 'times': ['03:30', '03:35', '04:00', '04:05', '04:15']},
    'user3': {'attempts': 8, 'times': ['22:15', '22:17', '22:20', '22:30', '22:35', '22:40', '23:00', '23:05']},
    'user4': {'attempts': 2, 'times': ['10:00', '10:15']},
    'user5': {'attempts': 1, 'times': ['11:00']}
}
threshold = 5

while True:
    for user, info in login_attempts.items():
        if info['attempts'] > threshold:
            print(f"사용자 {user}의 로그인 시도가 임계값을 초과했습니다.")
            print(f"로그인 시도 횟수: {info['attempts']}")
            print(f"로그인 시간: {info['times']}")
            break
    else:
        print("모든 사용자의 로그인 시도가 임계값 이하입니다.")
    break
````

### 문제 29
4명의 사용자가 참여하는 끝말잇기 게임을 프로그래밍으로 구현할고 합니다.

이 게임에서 참가자들은 차례대로 한 글자 이상의 단어를 입력하며, 각 단어는 이전 단어의 마지막 글자로 시작해야 합니다.

만약 규칙을 어기거나 참가자가 '종료'라고 입력할 경우, 게임이 종료됩니다. 게임이 종료되면 어떤 사용자가 게임을 종료시켰는지 알려주어야 합니다.

- 첫 번째 사용자로부터 게임의 첫 단어를 입력받습니다. 이 단어는 게임의 시작점이 됩니다.
- 이후, 사용자들은 순환적으로 단어를 입력해야 합니다. 순서는 1에서 4까지 반복됩니다.
- 입력된 단어는 이전 단어의 마지막 글자로 시작해야 하며, 한 글자인 단어는 허용되지 않습니다.
- 이미 입력된 단어를 다시 입력하는 것은 허용되지 않습니다.
- 사용자가 '종료'를 입력하거나 규칙을 어긴 경우, 해당 사용자를 출력하고 게임을 종료합니다.



예시 출력
```
끝말잇기 게임을 시작합니다. '종료'를 입력하면 게임이 종료됩니다.
첫 단어를 입력하세요: 사과
과로 시작하는 단어를 입력하세요: 과수원
원로 시작하는 단어를 입력하세요: 원숭이
이로 시작하는 단어를 입력하세요: 이사
사로 시작하는 단어를 입력하세요: 사과
User 1가 규칙에 맞지 않는 단어를 입력했습니다. 게임을 종료합니다.
```

### 정답

````python
print("끝말잇기 게임을 시작합니다. '종료'를 입력하면 게임이 종료됩니다.")

last_word = input("첫 단어를 입력하세요: ")
words = [last_word]
current_user = 1

while True:
    current_user = current_user % 4 + 1

    next_word = input(f"{last_word[-1]}로 시작하는 단어를 입력하세요: ")

    if next_word == '종료':
        break

    if len(next_word) == 1 or next_word in words or next_word[0] != last_word[-1]:
        print(f"User {current_user}가 규칙에 맞지 않는 단어를 입력했습니다. 게임을 종료합니다.")
        break

    words.append(next_word)
    last_word = next_word
````

### 문제 30
온라인 상점에서 물건을 구매하는 프로그램을 작성하려고 합니다. 사용자의 돈이 다 떨어지거나 사용자가 구매를 종료하기를 원할 때까지 구매를 계속합니다.

- 사용자가 초기에 가지고 있는 금액은 $2000입니다.
- 상점에서 판매하는 물품과 각 물품의 가격은 다음과 같습니다.

```
products = {'노트북': 1000, '스마트폰': 700, '헤드폰': 150, '책': 30}
```
- 사용자는 구매하고자 하는 물품의 이름을 입력합니다.
- 만약 입력된 물품이 판매 목록에 없다면, "판매 목록에 없습니다."라는 메시지를 출력해야 합니다.
- 사용자의 돈이 물품 가격보다 많거나 같아야 구매가 가능합니다.
- 구매 후 남은 돈을 갱신하고, 물품을 구매 목록에 추가합니다.
- 돈이 물품 가격보다 적으면, "돈이 부족해 구매할 수 없습니다."라는 메시지와 함께 구매를 중단합니다.
- 사용자가 "종료"를 입력하거나 돈이 $0에 도달하면, 구매를 중단하고 구매한 물품 목록과 남은 돈을 출력합니다.

예시출력

```
구매할 물건을 입력하세요 (구매 종료하려면 '종료' 입력):
스마트폰
스마트폰을(를) 구매했습니다. 남은 돈: $1300
구매할 물건을 입력하세요 (구매 종료하려면 '종료' 입력):
헤드폰
헤드폰을(를) 구매했습니다. 남은 돈: $1150
구매할 물건을 입력하세요 (구매 종료하려면 '종료' 입력):
책
책을(를) 구매했습니다. 남은 돈: $1120
구매할 물건을 입력하세요 (구매 종료하려면 '종료' 입력):
종료

구매한 물건 목록:
스마트폰
헤드폰
책
최종 남은 돈: $1120
```

### 정답

````python
products = {'노트북': 1000, '스마트폰': 700, '헤드폰': 150, '책': 30}

money = 2000

items = []

while True:
    print("구매할 물건을 입력하세요 (구매 종료하려면 '종료' 입력):")
    buy = input()
    if buy == '종료':
        break

    if buy in products:
        price = products[buy]

        if money >= price:
            money -= price
            items.append(buy)
            print(f"{buy}을(를) 구매했습니다. 남은 돈: ${money}")
        else:
            print(f"돈이 부족해 {buy}을(를) 구매할 수 없습니다. 남은 돈: ${money}")
            break
    else:
        print(f"{buy}은(는) 판매 목록에 없습니다.")

print("\n구매한 물건 목록:")

for item in items:
    print(item)

print(f"최종 남은 돈: ${money}")
````
