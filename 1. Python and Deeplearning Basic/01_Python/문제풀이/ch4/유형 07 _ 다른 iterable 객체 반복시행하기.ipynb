### 문제 26
string = 'abcdefg'일 때, 문자열 길이까지 2씩 증가시켜 아래와 같은 결과가 나오도록 코드를 작성해주세요.



```
AbCdEfG
```



### 정답

````python
string = 'abcdefg'

result_string = ''
for i in range(0, len(string), 2):
    result_string += string[i].upper()
    if i+1 < len(string):
        result_string += string[i+1]

print(result_string)
````

### 문제 30

```

    '일본': '후지산',

### 정답
