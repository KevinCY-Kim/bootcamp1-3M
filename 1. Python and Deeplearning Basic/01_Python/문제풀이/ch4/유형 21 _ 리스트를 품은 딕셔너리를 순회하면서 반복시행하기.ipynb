````python
user_tech_stacks = {
    'John': {'Python', 'JavaScript', 'HTML', 'CSS', 'Django', 'React'},
    'Alice': {'Java', 'JavaScript', 'Node.js', 'React', 'Vue.js'},
    'Bob': {'Python', 'Java', 'Spring', 'React', 'Angular'}
}

most_tech_stacks = 0
most_tech_user = None

for user1, tech_stacks1 in user_tech_stacks.items():
    if 'Python' not in tech_stacks1:
        continue

    current_tech_stacks = 0

    for user2, tech_stacks2 in user_tech_stacks.items():
        if user1 != user2:
            common_tech_stacks = tech_stacks1.intersection(tech_stacks2)
            current_tech_stacks = max(current_tech_stacks, len(common_tech_stacks))

    if current_tech_stacks > most_tech_stacks:
        most_tech_stacks = current_tech_stacks
        most_tech_user = user1

print(f"Python을 사용할 수 있으면서 가장 많은 기술 스택을 보유한 사람: {most_tech_user}")
````

### 정답

````python
movie_schedule = {
    'Nosferatu': ['14:00','16:10'],
    'His Girl Friday': ['16:30','19:00'],
    'The General': ['18:45','22:00'],
    'Metropolis': ['21:00','23:40']
}

min_time, max_time = '16:00', '20:00'

for movie, schedule in movie_schedule.items():
    is_within_range = False
    for time in schedule:
        if min_time <= time <= max_time:
            is_within_range = True
            break
    if is_within_range:
        print(f"Movie: {movie}")
        print(f"Total schedule: {len(schedule)}")
        for i, t in enumerate(schedule, start=1):
            if min_time <= t <= max_time:
                print(f"{i}. {t}")
        print("--------------------")
````

### 문제 29
아래의 menu와 allergies 정보를 고려하여 주문 가능한 메뉴의 재료를 확인하는 코드를 작성해보세요.
```
menu = {
    'hamburger': ['beef patty', 'lettuce', 'tomato', 'onion', 'pickles', 'ketchup', 'mayonnaise', 'bun'],
    'cheeseburger': ['beef patty', 'lettuce', 'tomato', 'onion', 'pickles', 'cheese', 'ketchup', 'mayonnaise', 'bun'],
    'french fries': ['potatoes', 'salt']
}

allergies = ['onion']
```

```
# 출력 결과

Item: cheeseburger
Orderable Ingredients: ['beef patty', 'lettuce', 'tomato', 'pickles', 'cheese', 'ketchup', 'mayonnaise', 'bun']
--------------------
Item: french fries
Orderable Ingredients: ['potatoes']
--------------------
```



### 정답

````python
menu = {
    'hamburger': ['beef patty', 'lettuce', 'tomato', 'onion', 'pickles', 'ketchup', 'mayonnaise', 'bun'],
    'cheeseburger': ['beef patty', 'lettuce', 'tomato', 'onion', 'pickles', 'cheese', 'ketchup', 'mayonnaise', 'bun'],
    'french fries': ['potatoes', 'salt']
}

allergies = ['onion']
order_menu = ['cheeseburger', 'french fries']

for item, ingredients in menu.items():
    if item in order_menu:
        orderable_ingredients = []

        if item == 'cheeseburger':
            for ingredient in ingredients:
                if ingredient not in allergies or ingredient != 'onion':
                    orderable_ingredients.append(ingredient)

        elif item == 'french fries':
            for ingredient in ingredients:
                if ingredient != 'salt':
                    orderable_ingredients.append(ingredient)

        print(f"Item: {item}")
        print("Orderable Ingredients:", orderable_ingredients)
        print("--------------------")
````
