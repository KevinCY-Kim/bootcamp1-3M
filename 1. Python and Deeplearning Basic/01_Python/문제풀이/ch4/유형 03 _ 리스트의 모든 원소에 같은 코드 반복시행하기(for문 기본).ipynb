### 문제 29
data 리스트에 있는 문자열에서 평당 가격이 가장 비싼 구의 이름을 출력하는 코드를 작성해보세요. (과학적 표기법은 'float'로 변환하세요.)

```python
      #   지역  평수  가격
data = ['강남구 20 1.00e+09',
        '서초구 25 1.20e+09',
        '송파구 30 1.80e+09']

max_price = 0
expensive_gu = ''

for datas in data:
    price = float(datas.split()[2]) / float(datas.split()[1])
    if price > max_price:
        max_price = price
        expensive_gu = datas.split()[0]

print(expensive_gu)
```

해당 국가명을 출력해보세요. (과학적 표기법은 'float'로 변환하세요.)
