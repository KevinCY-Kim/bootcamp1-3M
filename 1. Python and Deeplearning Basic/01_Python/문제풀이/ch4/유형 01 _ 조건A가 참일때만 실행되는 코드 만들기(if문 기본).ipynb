### 문제 29
img_height(높이), img_width(너비) = 110,275일 때, 이미지의 너비가 높이보다 크다면 이미지의 높이와 너비의 값을 서로 바꾸고

새롭게 이미지 사이즈를 224로 만들고자할 때 ratio라는 변수에 새로운 높이와 기존 높이 간의 비율을 계산하고

이미지의 새로운 너비와 높이를 계산하기 위해서 기존 너비와 높이에 ratio를 곱하고 정수형으로 변환하여

resize_img = (img_width, img_height) 튜플을 할당하는 코드를 작성해보세요.

````python
img_height, img_width = 100, 275

if img_width > img_height:
    img_height, img_width = img_width, img_height

img_size = 224
ratio = img_size / img_height
img_width = int(img_width * ratio)
img_height = int(img_height * ratio)
resize_img = (img_width, img_height)
resize_img
````

### 정답

````python
img_height, img_width = 110,275

if img_width > img_height:
    img_height, img_width = img_width, img_height

img_size = 224
ratio = img_size / img_height
img_width = int(img_width * ratio)
img_height = int(img_height * ratio)
resize_img = (img_width, img_height)
resize_img
````
