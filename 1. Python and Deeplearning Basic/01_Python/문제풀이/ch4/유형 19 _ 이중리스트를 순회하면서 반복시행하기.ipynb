### 문제 14
아래 코드를 읽고 최종적으로 생성된 grayscale_image 배열의 차원과 배열에 포함된 원소의 총 수를 맞추어 보세요.

---

<참고>

아래 코드는 이미지 데이터를 grayscale로 변환하는 프로세스를 구현한 코드입니다.

컬러 이미지는 주로 RGB (Red, Green, Blue) 색상 체계를 사용하여 표현되며


grayscale은 처리 효율성, 메모리 절약, 알고리즘 단순화 등의 이유로 사용됩니다.


````python
image = [
    [(100, 200, 50), (150, 75, 220), (80, 160, 40)],
    [(180, 200, 100), (150, 120, 90), (200, 180, 160)]
]

grayscale_image = []

for row in image:
    grayscale = []
    for pixel in row:
        grayscale_pixel = sum(pixel) // 3
        grayscale.append(grayscale_pixel)
    grayscale_image.append(grayscale)

print(grayscale_image)
````

이미지의 각 픽셀 값에 일정한 값을 더함으로써 전체 이미지를 밝게 만드는 가장 간단하고 직관적인 방법 중 하나입니다.

### 문제 16
아래 코드를 읽고 출력 결과를 맞춰보세요

---

<참고>

아래 코드는 이미지 데이터를 처리하여 특정 임계값(threshold) 이상의 픽셀 값만을 보존하고,

그 이하의 값을 가진 픽셀은 0으로 설정하는 임계값 필터링(thresholding) 과정을 수행하는 코드입니다.

이 과정은 임계값보다 낮은 노이즈를 0으로 설정함으로써 이미지에서 불필요한 정보를 줄이거나,

이미지의 데이터를 간소화하여 후속 처리를 용이하게 합니다.

````python
image = [[10, 20, 30], [40, 50, 60], [70, 80, 90]]
threshold = 50

filtered_image = []

for row in image:
    filtered = []
    for pixel in row:
        if pixel >= threshold:
            filtered.append(pixel)
        else:
            filtered.append(0)
    filtered_image.append(filtered)

print(filtered_image)
````

이 코드는 두 개의 리스트를 합치고, 각 학생의 평균 점수를 계산하여 새로운 리스트를 만드는 작업을 합니다. 기본적으로 nested_list에 있는 각 학생의 점수 리스트에 new_data_list에 있는 점수를 추가하고, 각 학생의 평균 점수를 계산하여 그 결과를 새로운 리스트에 저장합니다. 최종적으로 새로운 리스트를 출력합니다.

 - image 리스트에는 이미지의 픽셀 값이 포함되어 있습니다.
 - 첫 번째 for 루프에서는 image 리스트를 순회하면서 각 행을 처리합니다. inverted_row 빈 리스트를 만들어서 행의 반전된 픽셀 값을 저장합니다.
 - 두 번째 for 루프에서는 각 행을 순회하면서 픽셀 값을 반전시킵니다. 픽셀 값을 255에서 뺀 값을 inverted_pixel 변수에 저장하고 이 값을 inverted_row 리스트에 추가합니다.
 - 첫 번째 for 루프가 완료되면 inverted_image 리스트에 반전된 이미지 행을 추가합니다.


### 정답

````python
image = [[0, 255, 127], [64, 192, 128], [100, 50, 200]]

inverted_image = []

for row in image:
    inverted_row = []
    for pixel in row:
        inverted_pixel = 255 - pixel
        inverted_row.append(inverted_pixel)
    inverted_image.append(inverted_row)

print(inverted_image)
````

image_paths는 이미지 파일의 경로가 저장된 리스트이고, image_sizes는 해당 이미지 파일의 크기가 저장된 리스트입니다.

주어진 두 개의 리스트에서 파일 경로와 크기를 추출하여 새로운 리스트를 생성하는 코드를 작성해보세요.

```
image_paths = [['flower_01.jpg', 'flower_02.jpg'], ['flower_03.jpg', 'flower_04.jpg']]
image_sizes = [[(256, 256), (228, 228)], [(512, 512), (128, 128)]]

```


```
[['flower_01.jpg', (256, 256)], ['flower_02.jpg', (228, 228)], ['flower_03.jpg', (512, 512)], ['flower_04.jpg', (128, 128)]]
```



### 정답

````python
image_paths = [['flower_01.jpg', 'flower_02.jpg'], ['flower_03.jpg', 'flower_04.jpg']]
image_sizes = [[(256, 256), (228, 228)], [(512, 512), (128, 128)]]

results = []

for paths, sizes in zip(image_paths, image_sizes):
    for path, size in zip(paths, sizes):
        results.append([path, size])
print(results)
````
