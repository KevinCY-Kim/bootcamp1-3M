### 문제 15
아래 코드를 읽고 출력 결과를 맞춰보세요

````python
image = [
    [256, 256],
    [512, 512],
    [278, 355]
    ]
row, col = 0, 0
while row < len(image) and col < len(image[row]):
    #row, col 인덱스를 이용해서 2차원 배열을 순회, 범위에 벗어나지 않기 위함.
    image[row][col] = 255 - image[row][col]  #픽셀 값 반전 (255 => 0, 300 => -45 등)

#2차원 리스트를 wihile문으로 행(row)과 열(col)을 순회하기 위한 핵심 로직
    col += 1  #현재 행(row)의 다음 열(col) 로 이동, 같은행에서 옆으로 한칸 이동
    if col >= len(image[row]):  #열 인텍스가 현재 행에서 유효한 범위를 넘었는지 체크
        col = 0  #다음 행으로 넘어가기 전에 열 인덱스를 다시 0으로 초기화
        row += 1  #다음 행(row)으로 이동

for row in image: #최종 결과(변형된 이미지 데이터, 행렬 구조)를 한줄씩 출력하기 위함
    print(row)
````

### 문제 20

````python
numbers = 'onetwothreesevensix'

dic = {"one" : 1, "two" : 2, "three" : 3, "four" : 4, "five" : 5,
       "six" : 6, "seven" : 7, "eight" : 8, "nine" : 9, "zero" : 0 }

answer = 0
i = 0
j = 1
N = len(numbers)
while i<N:
    var = dic.get(numbers[i:j],False)
    if type(var) == int:
        answer *= 10
        answer += var
        i = j
        j += 1
    else:
        j += 1
answer
````

### 문제 30
회문이란 거꾸로 읽어도 제대로 읽는 것과 같은 문장이나 낱말, 숫자, 문자열을 의미합니다.


단어를 입력했을 때, 이 단어가 회문인지 아닌지 확인하는 코드를 생성하고자 합니다. 이와 같은 결과가 나오도록 해보세요.

* input()사용


```
단어를 입력하세요: racecar
입력한 단어는 거꾸로 읽어도 똑같은 단어입니다.
```



````python
word = input("단어를 입력하세요:")
resverse_word = ""

i = len(word) - 1
while i >= 0:
    reverse_word += word[i]
    i -= 1

if word == reverse_word:
    print("입력한 단어는 거꾸로 읽어도 똑같은 단어입니다.")
else:
    print("입력한 단어는 거꾸로 읽으면 다른 단어입니다.")
````

### 정답

````python
#  회문인지 확인하는 코드
# 'A man, a plan, a canal: Panama!'
# 'racecar', 'madam', 'level'
word = input("단어를 입력하세요: ")
reverse_word = ""

i = len(word) - 1
while i >= 0:
    reverse_word += word[i]
    i -= 1

if word == reverse_word:
    print("입력한 단어는 거꾸로 읽어도 똑같은 단어입니다.")
else:
    print("입력한 단어는 거꾸로 읽으면 다른 단어입니다.")
````
