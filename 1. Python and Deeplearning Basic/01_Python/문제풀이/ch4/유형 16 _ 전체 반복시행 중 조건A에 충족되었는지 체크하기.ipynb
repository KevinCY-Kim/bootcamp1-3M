### 문제 15
아래 코드를 읽고 출력 결과를 맞춰보세요

````python
news_articles = [
    "새로운 연구에 따르면, 운동은 건강에 이롭다고 합니다.",
    "지난 주말에 발생한 화재로 인해 많은 피해가 발생했습니다.",
    "정부는 코로나19 대응을 위해 추가적인 예방 조치를 발표했습니다."
]

keyword = "피해"
article_found = False

index = 0
while index < len(news_articles):
    if keyword in news_articles[index]:
        print("키워드 '{}'가 포함된 기사:".format(keyword))
        print(news_articles[index])

        start_index = 0
        while start_index < len(news_articles[index]):
            word_index = news_articles[index].find(keyword, start_index)
            if word_index != -1:
                context_start = max(0, word_index - 10)  # 키워드 앞쪽에서 10글자까지의 인덱스
                context_end = min(len(news_articles[index]), word_index + len(keyword) + 10)  # 키워드 뒤쪽에서 10글자까지의 인덱스
                print("앞뒤 문맥:", news_articles[index][context_start:context_end])
                start_index = word_index + len(keyword)
            else:
                break
        article_found = True
        break
    index += 1

if not article_found:
    print("해당 키워드를 포함한 기사가 없습니다.")
````

### 문제 23

limit = 50일 때, limit까지의 모든 소수를 출력하는 파이썬 프로그램을 작성하세요.

- 소수(prime number)는 1과 자기 자신 이외의 어떤 자연수로도 나누어 떨어지지 않는 양의 정수입니다.

````python
# 모든 소수를 출력하는 프로그램
# 소수는 1과 자기 자신 이외의 어떤 자연수로도 나누어 떨어지지 않는 양의 정수

for num in range(2,50+1):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
````

### 정답

````python
min_prime = None
max_prime = None

for num in range(2,100+1):
    is_prime = True
    for i in range(2, int(num ** 0.5) +1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        if min_prime is None:
            min_prime = num
        max_prime = num

if min_prime is None:
    print("소수를 찾지 못했습니다.")
else:
    print("가장 작은 소수:",  min_prime)
    print("가장 큰 소수:", max_prime)
````

```
log_data = {

    'user2': [('login', 8), ('logout', 7), ('view_page', 6), ('click_button', 4), ('send_email', 1)],
```

````python
log_data = {
    'user1': [('login', 10), ('logout', 8), ('view_page', 5), ('click_button', 3), ('download_file', 1)],
    'user2': [('login', 8), ('logout', 7), ('view_page', 6), ('click_button', 4), ('send_email', 1)],
    'user3': [('login', 12), ('logout', 9), ('view_page', 4), ('click_button', 2)],
    'user4': [('login', 15), ('logout', 12), ('view_page', 7), ('click_button', 5), ('download_file', 1)],
    'user5': [('login', 20), ('logout', 18), ('view_page', 10), ('click_button', 6)]
}

# 이벤트 당첨 여부를 나타내는 변수
event_winner_found = False

for user, events in log_data.items():
    login_count = 0
    for event, count in events:
        if event == 'login':
            login_count += count
            if login_count >= 15 and not event_winner_found:
                print(f"{user}님, 이벤트 당첨되었습니다!")
                event_winner_found = True
                break  # 이벤트 당첨된 사용자를 찾았으므로 더 이상 반복할 필요가 없음

# 이벤트 당첨자가 없는 경우 메시지 출력
if not event_winner_found:
    print("이벤트 당첨자가 없습니다.")
````

### 문제 30
당신은 로또 당첨 번호 맞추기 게임에 참가했습니다. 미리 정의된 당첨 번호가 있으며, 이 번호와 일치하는지 확인하는 게임입니다. 아래는 게임의 진행 방법입니다.

- 먼저, 미리 정의된 로또 당첨 번호 winning_numbers = [7, 14, 21, 28, 35, 42]가 있습니다.
- 사용자에게 1부터 45까지의 숫자 중 6개를 선택하도록 요청합니다. (공백으로 구분)
    - 사용자가 1 ~ 45 사이의 숫자를 입력하지 않거나, 6개의 숫자를 입력하지 않으면
    
    "잘못된 입력입니다. 1부터 45까지의 숫자 중 6개를 선택하세요."를 출력합니다.
- 사용자가 번호를 선택하면, 입력된 숫자와 당첨 번호를 비교하여 맞춘 번호의 개수를 확인하고 맞춘 개수를 출력합니다.
- 모든 번호를 맞춘 경우, 게임을 종료하고 "축하합니다! 모든 번호를 맞추셨습니다!"를 출력합니다.
- 하나의 번호라도 다른 경우, "아쉽지만 모든 번호를 맞추지 못했습니다. 다시 도전해보세요!"를 출력하고 게임을 계속 진행합니다.


출력 결과
```
로또 당첨 번호는 다음과 같습니다: [7, 14, 21, 28, 35, 42]
1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): 1 2 3 6 5
잘못된 입력입니다. 1부터 45까지의 숫자 중 6개를 선택하세요.
1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): 1 2 3 4 5 6
맞춘 번호 개수: 0
아쉽지만 모든 번호를 맞추지 못했습니다. 다시 도전해보세요!
1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): 7 14 21 28 35 42
맞춘 번호 개수: 6
축하합니다! 모든 번호를 맞추셨습니다!
```

### 정답

````python
winning_numbers = [7, 14, 21, 28, 35, 42]
print("로또 당첨 번호는 다음과 같습니다:", winning_numbers)

while True:
    user_input = input("1부터 45까지의 숫자 중 6개를 선택하세요(공백으로 구분): ")
    user_numbers = [int(num) for num in user_input.split()]

    if len(user_numbers) != 6 or any(num < 1 or num > 45 for num in user_numbers):
        print("잘못된 입력입니다. 1부터 45까지의 숫자 중 6개를 선택하세요.")
        continue

    matched_numbers = 0
    for number in user_numbers:
        if number in winning_numbers:
            matched_numbers += 1

    print("맞춘 번호 개수:", matched_numbers)

    if matched_numbers == 6:
        print("축하합니다! 모든 번호를 맞추셨습니다!")
        break
    else:
        print("아쉽지만 모든 번호를 맞추지 못했습니다. 다시 도전해보세요!")
````
