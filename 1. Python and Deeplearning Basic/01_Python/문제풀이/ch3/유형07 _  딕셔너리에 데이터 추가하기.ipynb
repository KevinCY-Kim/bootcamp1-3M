###문제 27




주어진 dataset 리스트에서 zip(), update()를 활용하여 아래와 같은 새로운 딕셔너리를 만들어주세요.  
- key 값 image는 dataset에 있는 모든 이미지 경로들을 value값으로 가집니다.  
- key 값 label는 dataset에서 각 이미지 경로가 가지는 key값을 value 값으로 가집니다.  
- 리스트 덧셈 연산과 곱셈 연산을 활용해주세요.


```
{'image': ['content/natural_images/airplane/airplane_0698.jpg',
            'content/natural_images/airplane/airplane_0563.jpg',
            'content/natural_images/car/car_0814.jpg',
            'content/natural_images/car/car_0800.jpg',
            'content/natural_images/car/car_0822.jpg',  
            'content/natural_images/dog/dog_0431.jpg',
            'content/natural_images/dog/dog_0253.jpg'],
'label': ['airplane', 'airplane', 'car', 'car', 'car', 'dog', 'dog']}

```




````python
dataset={'airplane': ['/content/natural_images/airplane/airplane_0698.jpg',
                      '/content/natural_images/airplane/airplane_0563.jpg'],

         'car': ['/content/natural_images/car/car_0814.jpg',
                 '/content/natural_images/car/car_0800.jpg',
                 '/content/natural_images/car/car_0822.jpg'],

         'dog': ['/content/natural_images/dog/dog_0431.jpg',
                 '/content/natural_images/dog/dog_0253.jpg']}
#주어진 dataset 리스트에서 zip(), update()를 활용하여 아래와 같은 새로운 딕셔너리
#key 값 image는 dataset에 있는 모든 이미지 경로들을 value값으로 가집니다.
#key 값 label는 dataset에서 각 이미지 경로가 가지는 key값을 value 값으로 가집니다.
#리스트 덧셈 연산과 곱셈 연산을 활용해주세요.

dataset_new = {}
key = ['image','label']
value = [dataset['airplane']+dataset['car']+dataset['dog'],
        ['airplane']*len(dataset['airplane'])+['car']*len(dataset['car'])+['dog']*len(dataset['dog'])]
dataset2 = zip(key, value)
dataset_new.update(dataset2)
dataset_new
````

````python
#@title 정답
dataset={'airplane': ['/content/natural_images/airplane/airplane_0698.jpg',
                      '/content/natural_images/airplane/airplane_0563.jpg'],

         'car': ['/content/natural_images/car/car_0814.jpg',
                 '/content/natural_images/car/car_0800.jpg',
                 '/content/natural_images/car/car_0822.jpg'],

         'dog': ['/content/natural_images/dog/dog_0431.jpg',
                 '/content/natural_images/dog/dog_0253.jpg']}


dataset2={}
key = ['image','label']
value = [dataset['airplane']+dataset['car']+dataset['dog'],
        ['airplane']*len(dataset['airplane'])+['car']*len(dataset['car'])+['dog']*len(dataset['dog'])]
new_data = zip(key, value)
dataset2.update(new_data)
print(dataset2)
````

###문제 30
주어진 products 딕셔너리의 'product1' 의 'stock'을 60으로 변경하고
새롭게 'product3': {'id': 'product3', 'name': 'Google Pixel 6', 'stock': 20}}을 추가해보세요.

````python
products = {
    'product1': {
        'id': 'product1',
        'name': 'Apple iPhone 14',
        'stock': 50
    },
    'product2': {
        'id': 'product2',
        'name': 'Samsung Galaxy S23',
        'stock': 30
    }
}
products['product1']['stock'] = 60
products_new = {'product3': {'id': 'product3', 'name': 'Google Pixel 6', 'stock': 20}}
products.update(products_new)
products
````

````python
#@title 정답
# 2차원 딕셔너리 수정&추가
products = {
    'product1': {
        'id': 'product1',
        'name': 'Apple iPhone 14',
        'stock': 50
    },
    'product2': {
        'id': 'product2',
        'name': 'Samsung Galaxy S23',
        'stock': 30
    }
}

products['product1'].update({'stock': 60})
products['product3'] = {'id': 'product3', 'name': 'Google Pixel 6', 'stock': 20}

print(products)
````
