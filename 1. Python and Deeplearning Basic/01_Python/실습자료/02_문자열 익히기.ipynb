### 문자열의 형변환

### 포맷 명세 미니 언어
- [fill]align][sign][#][0][width][grouping_option][.precision][type]
1. fill(채움) : 숫자, 일반문자, 특수문자 모두 가능
2. align(정렬) : <(왼쪽) / ^(가운데) / >(오른쪽)
3. width(너비) : 숫자로 명시
4. group(그룹) : ,
5. precision(정밀도) : .숫자f

```
# width
n1, n2, n3 = 1, 100, 10000

# 예시 : 이미지 파일명
txt1 = f'kakaotalk_{n1:5}.png'
txt2 = f'kakaotalk_{n2:5}.png'
txt3 = f'kakaotalk_{n3:5}.png'
print(txt1)
print(txt2)
print(txt3)
```

```
# align - width
n1 = 100

txt1 = f'kakaotalk_{n1:>5}.png'
txt2 = f'kakaotalk_{n1:^5}.png'
txt3 = f'kakaotalk_{n1:<5}.png'
print(txt1)
print(txt2)
print(txt3)
```

```
# fill - width
label = 'cat'
num1, num2, num3 = 1, 10, 100

filepath1 = f'{label}_{num1:03}.jpg'
filepath2 = f'{label}_{num2:03}.jpg'
filepath3 = f'{label}_{num3:03}.jpg'
print(filepath1)
print(filepath2)
print(filepath3)
```

```
# fill - align - width
# 1
label = 'cat'
num1 = 3

filepath1 = f'{label}_{num1:0>3}.jpg'
filepath2 = f'{label}_{num1:0^3}.jpg'
filepath3 = f'{label}_{num1:0<3}.jpg'
print(filepath1)
print(filepath2)
print(filepath3)
print('-'*30)

# 2
filepath1 = f'{label}_{num1:*>3}.jpg'
filepath2 = f'{label}_{num1:*^3}.jpg'
filepath3 = f'{label}_{num1:*<3}.jpg'
print(filepath1)
print(filepath2)
print(filepath3)
```

```
# 문자 : fill - align - with
item = 'coat'

item_desc = f'label:{item}'
print(item_desc)

item_desc1 = f'label:{item:*<10}'
print(item_desc1)

item_desc2 = f'label:{item:*^10}'
print(item_desc2)
item_desc3 = f'label:{item:*>10}'
print(item_desc3)
```

```
# group
price = 1000000

txt = f'가격 : {price:,}'
print(txt)
```

```
# fill-align-width-group
price = 1000000

txt = f'가격 : {price:*<11,}'
print(txt)
```

```
# precision
prob = 0.811111111

txt = f'확률 : {prob:.3f}'
print(txt)
```

```
# fill-align-width-precision
prob = 0.811111111

txt = f'확률 : {prob:*^10.3f}'
print(txt)
```
