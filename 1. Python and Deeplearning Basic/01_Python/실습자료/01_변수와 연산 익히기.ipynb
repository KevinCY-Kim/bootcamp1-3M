### 비교연산자/논리연산자

```
# 5가 3보다 크다
5 > 3
```

```
# 5가 3보다 작다
5 < 3
```

```
# 5가 3보다 크거나 같다
5 >= 3 # 5 => 3
```

```
# 5가 3보다 작거나 같다
5 <= 3
```

```
# 5가 3과 같다
5 == 3
```

```
# 5가 3과 같지 않다
5 != 3
```

```
# 5가 3보다 크고, 4가 2보다 크다
print(5 > 3 and 4 > 2)

# 5가 3보다 크고, 4가 2보다 작다
print(5 > 3 and 4 < 2)
```

and : 둘 다 참이어야 참 (둘 중 하나라도 거짓이면 거짓)

```
# 4가 2보다 크거나, 4가 8보다 크다
print(4 > 2 or 4 > 8)

# 4가 2보다 작거나, 4가 8보다 크다
print(4 < 2 or 4 > 8)
```

or : 둘 중 하나만 참이어도 참 (둘 다 거짓이어야 거짓)

변수로 표현하기

```
# a가 0보다 크다/크거나 같다/작다/작거나 같다
a = 5
a > 0, a >= 0, a < 0, a <= 0
```

```
# a가 b와 같다/ 같지 않다
a = 5
b = 5

a == b, a != b
```

```
a, b, c = 4, 2, 8

# a가 b보다 크고, a가 c보다 크다
print(a > b and a > c)

# a가 b보다 크고, a가 c보다 작다
print(a > b and a < c)

# a가 b보다 크거나, a가 c보다 크다
print(a > b or a > c)

# a가 b보다 작거나, a가 c보다 크다
print(a < b or a > c)

# a가 b보다 작지 않다
print(not a < b)
```

```
a, b, c = 4, 2, 8
print(a > b and a > c)
print(a > b or a > c)
not a < b # 마지막 줄 객체는 출력하도록 설정되어 있다.
```

## 기본 내장 연산 함수 익히기
- abs(-5) : 절대값
- round(5.114, 2) : 반올림
- divmod(5, 3) : 몫과 나머지 한번에
- pow(5, 3) : 제곱
- min(1, 2, 3) : 최소값
- max(1, 2, 3) : 최대값

```
# 절대값 계산하기
a, b, c = -6, -5, -1
abs(a), abs(b), abs(c)
```

```
# 반올림하기
a, b = 3.6, 3.45
round(a), round(b)
```

```
c = 3.456
round(c), round(c, 1), round(c, 2)
```

```
d = 3456
round(d), round(d, -1), round(d, -2)
```

```
# 몫과 나머지 한번에 연산하기
a, b = 5, 3

result = divmod(a,b)
print(result)
```

```
# 제곱하기
a, b, c, d = 2, 4, 10, 3

num1 = pow(a, b)
num2 = pow(c, d)
print(num1, num2)
```

```
# 최소값, 최대값 연산하기
a,b,c,d = 2,4,3,5

print(max(a,b,c,d))
print(min(a,b,c,d))
```

## 변수의 타입 변환 익히기

- 변수 타입의 종류
1. 정수 (int)
2. 실수 (float)
3. 문자열 (str)
4. 논리값 (bool)

- 가능한 경우에 한하여, 서로 다른 타입으로 변환이 가능하다.


```
# 변수 타입의 종류
a = 10
print(a, type(a))

b = 0.1
print(b, type(b))

c = '파이썬'
print(c, type(c))

d = True
print(d, type(d))

e = False
print(e, type(e))
```

```
# 실수를 정수로
a = 3.7

result = int(a)
print(type(result), result)
```

```
a, type(a) # 형변환은 원본 보존한다.
```

```
# 정수를 실수로
a = 13

result = float(a)
print(type(result), result)
```

```
# 숫자를 문자열로
num1 = 123
num2 = 0.123

result1 = str(num1) # '123'
result2 = str(num2)
print(type(result1), result1)
print(type(result2), result2)
```

```
# 문자열을 숫자로
str1 = '123'
str2 = '456'

result1 = int(str1)
result2 = float(str2)
print(type(result1), result1)
print(type(result2), result2)
```

```
# 변환 불가한 경우
str3 = '파이썬'

result = int(str3)
# result = float(str3)

print(type(result), result)
```

```
# 숫자를 논리값으로
num1 = 10
num2 = 0 # 존재x 의미 --> 거짓

result1, result2 = bool(num1), bool(num2)
print(type(result1), result1)
print(type(result2), result2)
```

```
# 문자열를 논리값으로
num1 = '0'
num2 = '' # null

result1, result2 = bool(num1), bool(num2)
print(type(result1), result1)
print(type(result2), result2)
```

```
# 논리값을 숫자로
t, f = True, False

n1, n2 = int(t), float(f)
print(type(n1), n1)
print(type(n2), n2)
```

```
# 논리값을 문자열로
t, f = True, False

s1, s2 = str(t), str(f)
print(type(s1), s1) # 'True'
print(type(s2), s2) # 'False'
```
