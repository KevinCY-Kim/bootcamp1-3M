#### 함수가 품고 있는 코드를 변경해서 사용하고 싶다면?

````python
# 10배라고 고정되어있던 값을 원하는 값으로 수정하여 사용가능하도록 함수로 선언해보자.
hangles = '가 나 다 라 마 바 사'.split()

han2idx = {hangle : idx for idx, hangle in enumerate(hangles)}
han2idx

for key in han2idx:
    han2idx[key] *= 10 # factor 변수로 수정
print(han2idx)
````

````python
def sample_code(factor): # factor 변수를 함수명 뒤 괄호 안에 명시
    hangles = '가 나 다 라 마 바 사'.split()

    han2idx = {hangle : idx for idx, hangle in enumerate(hangles)}
    han2idx

    for key in han2idx:
        han2idx[key] *= factor # 10이라고 고정되어있던 값을 factor 변수로 수정

    print(han2idx)
````

````python
# 함수를 호출할 때 factor 변수에 원하는 값을 선언할 수 있음
sample_code(10) # factor = 10
sample_code(100) # factor = 100
sample_code(300) # factor = 300
````

### Default값 설정

#### 함수를 호출할 때, 함수의 입력 인자 개수를 맞춰주지 않으면 어떻게 될까?

create_info_dict의 input인자는 3개다.  
만약 input인자를 2개만 넘겨주면 어떻게 될까?  

````python
def create_info_dict(name, phone, email):
    data = {'name': name,
           'phone-number': phone,
           'email': email}
    return data
````

````python
create_info_dict('dino', '010-****-****')
````

type error 발생!  
- email 경우에 안넣어도 에러 없이 수행되도록 만들고 싶다면?
    - email에 default값을 설정해주면 됨


````python
# email에 None이라는 default 값 설정하기
def create_info_dict(name, phone, email=None):
    data = {'name': name,
           'phone-number': phone,
           'email': email}
    return data
````

````python
create_info_dict('dino', '010-****-****') # default 설정을 해둔 인자는 할당해주지 않아도 괜찮다.
````

#### 문제 2
숫자 여러개를 입력받아 짝수를 필터링하여 반환하는 even_num함수를 작성해보세요.
- 입력인자 : args (가변인자로 받을 것)
- return : 짝수만 필터링한 리스트
- 짝수필터링 : 내포용법을 사용하여 작성할 것

````python
# @title 정답
def even_num(*args):
    even_numbers = [num for num in args if num % 2 == 0]
    return even_numbers

result = even_num(1, 2, 3, 4, 5, 6)
print(result)

result = even_num(10, 15, 20, 25, 30)
print(result)
````

### 람다함수 (별명 : 간단함수 혹은 한줄함수)
def 함수 선언 문법이 과하게 느껴질만큼  
간단한 코드(한줄 코드)만 함수화 하고싶은 경우에 사용되는 문법

````python
# 위의 예제 코드만 봐도 딱 한줄로 축약 가능
def even_num(*args):
    return [num for num in args if num % 2 == 0]
````

함수로 선언하고자 하는 코드가 한줄 뿐이라면,  
등호 써도 되는 거 아닐까?

````python
even_num(1, 2, 3, 4, 5)
````

[본능 주입시키기]
- 한줄짜리 코드를 함수화하고 싶을 때
    - "람다마렵다."


````python
# 람다 선언하는 방법 : 함수명 = 램브다 입력 : 출력
even_num = lambda *args: [num for num in args if num % 2 == 0]

even_num(1, 2, 3, 4, 5)
````

````python
# x를 입력받아 제곱(square)한 값을 리턴해주는 함수
def mySquare(x):
    return x**2
````

주어진 파일명 리스트 filenames와 확장자 extension가 주어졌을 때, 해당 확장자를 가진 파일명만 필터링하여 출력하는 파이썬 스크립트를 작성하세요.

이 작업을 수행하기 위해 람다 함수를 사용하고, 리스트 컴프리헨션을 이용하세요.

- extension: 필터링할 파일의 확장자 (예: 'jpg')
```
filenames = ['readme.txt', 'car.jpg', 'airplane.jpg', 'person.png', 'flower.jpg', 'bicycle.bmp', 'cat.jpg']
```



##### 정답

````python
filenames = ['readme.txt', 'car.jpg', 'airplane.jpg', 'person.png', 'flower.jpg', 'bicycle.bmp', 'cat.jpg']
extension = 'jpg'

files_with_ext = lambda fnames, ext : [fname for fname in fnames if fname[-3:] == ext]
files_with_ext(filenames, extension)
````

````python
# 문자열 메서드 endswith / startswith
files_with_ext = lambda fnames, ext: print([fname for fname in fnames if fname.endswith(ext)])

filenames = ['readme.txt', 'car.jpg', 'airplane.jpg', 'person.png', 'flower.jpg', 'bicycle.bmp', 'cat.jpg']
extension = 'jpg'

files_with_ext(filenames, extension)
````

### 지역변수와 전역변수
- 지역변수 : 함수 내에서 선언된 변수
- 전역변수 : 함수 밖에서 선언된 변수  


전역변수명과 지역변수명이 중복되어도 정말 괜찮을까?

````python
# 모든 변수명을 a로 선언해보자.
a = 5 # 전역변수

def fun1():
  a = 1 # 지역변수 1
  print(a) # -> 이 a 는 누구일까?

def fun2():
  a = 2 # 지역변수 2
  print(a) # -> 이 a 는 누구일까?

def fun3():
  print(a) # -> 이 a 는 누구일까?

def fun4(): # 전역변수를 함수 안에서 수정까지 하려면
  global a # 전역변수 a를 끌어온다.
  a = 4 # 전역변수
  print(a) # 이 a 는 누구일까?
````

1. 지역변수가 있다면, 함수 내에서는 지역변수를 우선한다.
2. 지역변수가 없다면, 함수 내에서도 전역변수를 참조한다.
3. 함수내에서 전역변수를 호출하여 값을 수정할 수 있다.

````python
print(a) # -> 전역변수 a
````

````python
# 여기서 다시 a를 출력하면 값이 무엇일까?
print(a) # -> 전역변수 a = 5
````
