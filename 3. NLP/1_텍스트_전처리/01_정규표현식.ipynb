# 2. 정규 표현식(Regular Expression)

## 2.1. 정규 표현식 패턴 정의하기

1. \[ \] : 괄호 안 모든 알파벳
    1. [^] : ~을 제외하고
    1. [a-z] : a~z까지 모든 알파벳  
1. *, +, \{숫자\} : 반복되는 문자열
1. ? : 있거나 없거나
1. . : 모든 문자열
    1. \d, \D : digit 숫자
    1. \w, \W : word 문자
    1. \s, \S : space 공백
1. greedy vs reluctant
1. \( \) : 패턴에서 원하는 범위만 선택

### 몇 번이나 나올까?

앞에서 [  ]만으로는 너무 낱개로 잘라져서 나오죠?..

정규표현식에는 해당 패턴이 몇 번 나오는지 정할 수 있는 몇 가지 방법이 있습니다.

#### ? - 있느냐 없느냐

해당 글자가 있든 없든 매칭시킵니다.

```python
string = 'apple app aple apppple a00e aqqle ale APPLE Apple (apple) [apple]'
re.findall('ap?le', string) # 물음표 앞에 p가 있거나 혹은 없는 경우 모두
```

#### {} - 몇 번 반복되느냐

{n} 또는 {n, m}은 몇 번 반복되는지 지정할 수 있습니다.

```python
re.findall('ap{2}le', string)  # 중괄호는 p가 두번 반복되는 경우만
```

```python
re.findall('ap{2,5}le', string)  #중괄호 안에 숫자 범위 p가 2~5번 반복되는 경우만
```

#### + - 1번 이상 반복되느냐?

+를 쓰면 1번 이상 반복되는 패턴을 찾을 수 있습니다.

```python
re.findall('ap+le', string)
```

#### * - 0번 이상 반복되느냐?

*는 +와 유사하지만 없는 경우, 즉 0번 이상 반복패턴을 찾을 수 있습니다.

```python
re.findall('ap*le', string)
```

### 특수한 기능을 하는 문자 클래스 - escape code

문자 클래스는 공백, 숫자, 글자등을 대표하는 표기법입니다.

- \d (digit) 숫자
- \D 숫자가 아닌거


- \s (space) 공백 <-- 여기서 '공백'이란 스페이스, 탭, 줄바꿈을 모두 포함
- \S 공백이 아닌거


- \w (word) 글자
- \W 글자가 아닌거


- \t tap
- \\\ 단어로서의 백슬러시
- \n 줄 바꿈(new line)

## 2.2 응용하기


```python
example = 'class는 attribute와 method로 이루어져 있습니다.'
print(re.findall('[가-힣]+', example))
print(re.findall('[a-zA-Z]+', example))
print(re.findall('\w+', example))
```

```python
# phone = "Phone: +1 (123) 4546-7890"
phone = "Phone: +82 (055) 4563-7890"
# phone = "Phone: +82 (02) 3449-7890"

print(re.findall('[0-9]+', phone))  #1~5까지 숫자를 찾아줘
print(re.findall('\+\d+', phone))  #+가 앞에 붙은 숫자를 찾아줘
```

### Greedy 욕심쟁이 VS Reluctant(Lazy) 소심이 --> ?

다음과 같은 예시를 한 번 보자

```python
text = "<b>Hello</b><b>World</b>"
print(re.findall(r'<b>.*</b>', text))
print(re.findall(r'<b>.*?</b>', text))
```

위의 경우 가장 앞에 있는 \<b>부터 가장 뒤에 있는 \</b>사이에 모든 단어가 한번에 parsing됩니다. (greedy)

실제로 내가 원하는 것은 html에 있는 각 tag 속의 element이므로 더 작은 단위로 잘리길 원합니다.

그럴경우 다음과 같이 __?__를 붙여주면 해당하는 조건에 맞는 가장 작은 단위로 잘라줍니다(reluctant/lazy)

```python
re.findall(r'<b>.*?</b>', text)
```

### 그루핑 - 내가 원하는 일부만

패턴으로 찾았지만 뽑고 싶은건 그 중 일부만일 수 있습니다.

(  )를 사용하면 패턴 중에서 원하는 범위만 추출할 수 있습니다.

```python
re.findall(r'<b>.*?</b>', text) # 이 경우 출력 item의 앞 뒤에 tag가 그대로 붙어있다.
```

```python
re.findall(r'<b>(.*?)</b>', text)  #element만 필요하기 때문에 그루핑을 통해서 추출한다.
```

주의!!

기본 괄호\(\)는 그루핑의 용도로 사용되기 때문에 문장 내에서 괄호를 표현하기 위해서는 excape code를 사용해야 합니다.

괄호 앞에 \\가 있는지 없는지 잘 봅시다.

```python
aca = "솔데스크의 SOL은 Solomon왕의 지혜처럼 우리의 인생을 성공하기 위해(Success of Life) 다양한 해결방안(Solution)을 제공하는 회사가 되길 바라는 의미입니다."

re.findall('\(.+?\)', aca)
# print(re.findall('\((.+)\)', aca))  #괄호 사이에 있는 모든 text를 구하고 싶다. - greedy
# print(re.findall('\((.+?)\)', aca))  #괄호 사이에 있는 모든 text를 구하고 싶다. - lazy
```

```python
phone = "Phone: +82 (055) 4563-7890, +82 (02) 1111-2376"
print(re.findall('\+(\d+)', phone))  # 국가 코드
print(re.findall('\((\d+)\)', phone))  # 지역 번호
print(re.findall('(\d+-\d+)', phone)) # 전화 번호
```

#### 그룹 인덱스
그룹에 여러개일 때 \1, \2와 같이 백슬러시 뒤에 숫자를 적으면 해당 순서의 그룹을 지정할 수 있다.

```python
text = '961204-1824341'
#print(re.sub('(\d+)-(\d+)', r'그룹 1 : \1, 그룹 2 : \2', text))
re.findall('(\d+)-(\d+)', text)
new_text = re.sub(r'(\d+)-(\d+)', r'\1-\2', text)
print(new_text)
```

```python
text = '고객님께서 입력하신 주민등록 번호는 961204-1824341 입니다.'

print(re.sub('(\d{6}-\d)(\d{6})', '\1*', text))  # \1은 escape 시쿼스로 '아스키 코드'를 나타냅니다.
print(re.sub('(\d{6}-\d)(\d{6})', '\\1*', text))  # 백슬러시로 인식합니다.
print(re.sub('(\d{6}-\d)(\d{6})', r'\1*', text)) #raw string을 쓰는게 명확하게 표현할 수 있으므로 권장!!

print()
print(re.sub('(\d{6}-\d)(\d{6})', r'*\2', text))
```

```python
text = '고객님께서 입력하신 주민등록 번호는 961204-1824341 입니다.'
new_text = re.sub('(\d{6}-\d)(\d{6})', r'\1******', text)
print(new_text)
```

```python
text = "오늘은 2025-06-27이고, 내일은 2025-06-28입니다."
new_text = re.sub(r'(\d{4})-(\d{2})-(\d{2})', r'\1년\2월\3일', text)
print(new_text)
```

```python
text = "오늘은 2025-06-27이고, 내일은 2025-06-28입니다."

print(re.sub(r'(\d{d4})-(\d{2})-(\d{2})', r'\1년 \2월 \3일', text))
```
