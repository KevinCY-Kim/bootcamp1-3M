### (2) ê¸°ë³¸ ì±—ë´‡ ì‘ë‹µ ìƒì„± ê³¼ì •


- íŒŒì¸íŠœë‹ ì „ì— ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ì˜ ê¸°ë³¸ ì„±ëŠ¥ì„ í™•ì¸í•©ë‹ˆë‹¤.
- ì‚¬ìš©ì ì§ˆë¬¸ì— ëŒ€í•œ ì‘ë‹µì„ generate()ë¡œ ì§ì ‘ ìƒì„±í•´ë³´ê² ìŠµë‹ˆë‹¤.

- model.generate : ëª¨ë¸ì´ ì‘ë‹µì„ ìƒì„±í•  ë•Œ í’ˆì§ˆì„ ì¡°ì ˆí•˜ëŠ” ë‹¤ì–‘í•œ í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ ì„¤ì •í•©ë‹ˆë‹¤.
    - ì£¼ìš” íŒŒë¼ë¯¸í„°
        - temperature : ì˜ˆì¸¡ê°’ì— ë¬´ì‘ìœ„ì„± ë¶€ì—¬, ë‚®ì„ìˆ˜ë¡ ì¼ê´€ì ì¸ ì‘ë‹µ, ë†’ì„ìˆ˜ë¡ ì°½ì˜ì  ì‘ë‹µ.
        - top_p : ëˆ„ì  í™•ë¥  ê¸°ì¤€ìœ¼ë¡œ ìƒìœ„ í† í°ë§Œ ìƒ˜í”Œë§. ë‚®ì„ìˆ˜ë¡ ì•ˆì •ì , ë†’ì„ìˆ˜ë¡ ë‹¤ì–‘í•¨.
        - repetition_penalty : ê°™ì€ ë¬¸ì¥ì´ë‚˜ ë‹¨ì–´ ë°˜ë³µì„ ì–µì œ. ìì—°ìŠ¤ëŸ¬ìš´ ì‘ë‹µ ìœ ë„.
        - do_sample : ìƒ˜í”Œë§ ë°©ì‹ì„ ì ìš©í•˜ì—¬ ë‹¤ì–‘í•œ ë‹µë³€ ìƒì„± ê°€ëŠ¥ (Trueì¼ ê²½ìš° ëœë¤ì„± ë¶€ì—¬).
        - max_new_tokens : ìƒì„±ë˜ëŠ” ì‘ë‹µì˜ ìµœëŒ€ í† í° ìˆ˜ë¥¼ ì œí•œ (ì‘ë‹µ ê¸¸ì´ ì œì–´).

````python
user_input = "ìµœì €ì„ê¸ˆì œë€ ë¬´ì—‡ì¸ê°€ìš”?"

# âœ… ì…ë ¥ì„ ëª¨ë¸ì´ ì²˜ë¦¬í•  ìˆ˜ ìˆëŠ” í…ì„œë¡œ ë³€í™˜
inputs = tokenizer(user_input, return_tensors="pt").to("cuda")
print(f"ğŸ“ ì§ˆë¬¸: {user_input}")

# âœ… ëª¨ë¸ ì‘ë‹µ ìƒì„± (ì¶”ë¡  ì˜µì…˜ ì ìš©)
outputs = model.generate(
    **inputs,
    max_new_tokens=100,        # ì‘ë‹µ ìµœëŒ€ ê¸¸ì´ ì œí•œ
    temperature=0.5,           # ì˜¨ë„ ì¡°ì ˆì„ í†µí•´ ì¼ê´€ëœ ì‘ë‹µ ìœ ë„
    top_p=0.85,                # ìƒìœ„ í™•ë¥  í† í°ë§Œ ìƒ˜í”Œë§
    repetition_penalty=1.2,    # ë°˜ë³µ ë°©ì§€
    do_sample=True             # ìƒ˜í”Œë§ ë°©ì‹ ì ìš©
)

# âœ… ì‘ë‹µ ë””ì½”ë”©
decoded_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("ğŸ¤– ì‘ë‹µ:", decoded_output)
````

ğŸ§¾ Alpaca-style ë°ì´í„° í¬ë§·ì´ë€?

Alpaca-style í¬ë§·ì€ LLMì„ Instruction ê¸°ë°˜ìœ¼ë¡œ íŒŒì¸íŠœë‹í•  ë•Œ ì‚¬ìš©í•˜ëŠ” ëŒ€í‘œì ì¸ í˜•ì‹ì…ë‹ˆë‹¤.
ëª¨ë¸ì´ ì–´ë–¤ ëª…ë ¹ì„ ìˆ˜í–‰í•´ì•¼ í•˜ëŠ”ì§€ ëª…í™•íˆ ì•Œë ¤ì£¼ëŠ” êµ¬ì¡°ë¡œ, íŠ¹íˆ ì •ë³´ ì œê³µí˜• ë‹¨ì¼í„´ ì±—ë´‡ì— ì í•©í•©ë‹ˆë‹¤.


- í˜•ì‹
```
    Below is an instruction that describes a task. Write a response that appropriately completes the request.

    ### Instruction:
    {instruction}

    ### Response:
    {output}<|end_of_text|>
```

````python
from datasets import load_dataset

# âœ… ë°ì´í„°ì…‹ ë¡œë“œ
dataset = load_dataset("jihye-moon/LawQA-Ko", split="train")

# âœ… EOS í† í°
EOS_TOKEN = tokenizer.eos_token

# âœ… Alpaca-style í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿
alpaca_prompt = """Below is an instruction that describes a task. Write a response that appropriately completes the request.

                ### Instruction:
                {}

                ### Response:
                {}"""
````

````python
# âœ… Step 1: instruction / output í•„ë“œ ìƒì„±
def to_alpaca_format(example):
    return {
        "instruction": example["question"],
        "output": example["answer"]
    }

dataset = dataset.map(to_alpaca_format)
````

output ê¸¸ì´ê°€ ë„ˆë¬´ ê¸´ ê²½ìš° í•™ìŠµì´ ì œëŒ€ë¡œ ë˜ì§€ ì•Šì„ ìˆ˜ ìˆê¸° ë•Œë¬¸ì—, outputì˜ í† í° ìˆ˜ê°€ 516 ì´í•˜ì¸ ìƒ˜í”Œë§Œ ì‚¬ìš©í•˜ê² ìŠµë‹ˆë‹¤.

````python
# âœ… Step 2: output ê¸¸ì´ í•„í„°ë§ (í† í° ê¸¸ì´ 516 ì´í•˜)
def is_output_short(example):
    tokenized = tokenizer(example["output"], truncation=False, add_special_tokens=False)
    return len(tokenized["input_ids"]) <= 516

filtered_dataset = dataset.filter(is_output_short) # Trueì¸ ìš”ì†Œë§Œ ì„ ë³„
````

````python
# âœ… Step 3: text í•„ë“œ ìƒì„±
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    outputs = examples["output"]
    texts = []
    for instruction, output in zip(instructions, outputs):
        text = alpaca_prompt.format(instruction, output) + EOS_TOKEN
        texts.append(text)
    return { "text": texts }

formatted_dataset = filtered_dataset.map(formatting_prompts_func, batched=True)

# âœ… ê²°ê³¼ í™•ì¸
print(f"ğŸ“Š í•„í„°ë§ í›„ ìƒ˜í”Œ ìˆ˜: {len(formatted_dataset)}")
print(formatted_dataset[0])
````

````python
from unsloth import FastLanguageModel
from transformers import TextStreamer
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "MLP-KTLim/llama-3-Korean-Bllossom-8B",  # âœ… ì‚¬ìš© ëª¨ë¸
    max_seq_length = 2048,      # âœ… ê¶Œì¥ ì„¤ì •
    dtype = None,               # âœ… float16 or bfloat16 ìë™ ì„ íƒ
    load_in_4bit = True         # âœ… 4bit QLoRA ì–‘ìí™” ë¡œë”© í™œì„±í™”
)
````

### (3) LoRA(QLoRA)ë¥¼ ì´ìš©í•œ ëª¨ë¸ íŒŒì¸íŠœë‹

- GPT-4 ë˜ëŠ” BERT ê¸°ë°˜ í‰ê°€ ëª¨ë¸ì„ í™œìš©í•˜ì—¬ ì±—ë´‡ ì‘ë‹µì˜ ìì—°ìŠ¤ëŸ¬ì›€ê³¼ ì˜ë¯¸ì  ìœ ì‚¬ë„ë¥¼ ë¹„êµ.

````python
from nltk.translate.bleu_score import sentence_bleu
from rouge import Rouge
import nltk
nltk.download('punkt')

eval_size = 10

bleu1_list, bleu2_list = [], []
rouge1_list, rouge2_list, rougel_list = [], [], []
distinct1_list, distinct2_list = [], []

rouge = Rouge()

for i in range(eval_size):
    reference = formatted_dataset[i]["output"].strip()
    instruction = formatted_dataset[i]["instruction"].strip()

    prompt = f"""Below is an instruction that describes a task. Write a response that appropriately completes the request.

### Instruction:
{instruction}

### Response:
"""

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(
        **inputs,
        max_new_tokens=512,
        do_sample=True,
        temperature=0.7,
        top_p=0.95,
        eos_token_id=tokenizer.eos_token_id,
    )
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

    if "### Response:" in decoded:
        generated = decoded.split("### Response:")[-1].strip()
    else:
        generated = decoded.strip()

    try:
        bleu1 = sentence_bleu([reference.split()], generated.split(), weights=(1, 0, 0, 0))
        bleu2 = sentence_bleu([reference.split()], generated.split(), weights=(0.5, 0.5, 0, 0))
        rouge_scores = rouge.get_scores(generated, reference)[0]
        tokens = generated.split()
        bigrams = list(zip(tokens, tokens[1:]))
        distinct1 = len(set(tokens)) / max(len(tokens), 1)
        distinct2 = len(set(bigrams)) / max(len(bigrams), 1)

        bleu1_list.append(bleu1)
        bleu2_list.append(bleu2)
        rouge1_list.append(rouge_scores["rouge-1"]["f"])
        rouge2_list.append(rouge_scores["rouge-2"]["f"])
        rougel_list.append(rouge_scores["rouge-l"]["f"])
        distinct1_list.append(distinct1)
        distinct2_list.append(distinct2)
    except:
        print(f"âš ï¸ {i}ë²ˆì§¸ ìƒ˜í”Œ í‰ê°€ ì‹¤íŒ¨")

print("ğŸ“Š [1000ê°œ ìƒ˜í”Œ í‰ê·  í‰ê°€ ê²°ê³¼]")
print(f"BLEU-1 í‰ê· : {sum(bleu1_list)/len(bleu1_list):.4f}")
print(f"BLEU-2 í‰ê· : {sum(bleu2_list)/len(bleu2_list):.4f}")
print('-------------------------')
print(f"ROUGE-1 í‰ê· : {sum(rouge1_list)/len(rouge1_list):.4f}")
print(f"ROUGE-2 í‰ê· : {sum(rouge2_list)/len(rouge2_list):.4f}")
print(f"ROUGE-L í‰ê· : {sum(rougel_list)/len(rougel_list):.4f}")
print('-------------------------')
print(f"Distinct-1 í‰ê· : {sum(distinct1_list)/len(distinct1_list):.4f}")
print(f"Distinct-2 í‰ê· : {sum(distinct2_list)/len(distinct2_list):.4f}")
````

````python
from unsloth import FastLanguageModel
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "MLP-KTLim/llama-3-Korean-Bllossom-8B",  # âœ… ì‚¬ìš© ëª¨ë¸
    max_seq_length = 2048,      # âœ… ê¶Œì¥ ì„¤ì •
    dtype = None,               # âœ… float16 or bfloat16 ìë™ ì„ íƒ
    load_in_4bit = True         # âœ… 4bit QLoRA ì–‘ìí™” ë¡œë”© ë¹„í™œì„±í™”
)


model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    lora_alpha = 16,
    lora_dropout = 0.0,
    bias = "none",
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    use_gradient_checkpointing = "unsloth",
    use_dora = True,  # âœ… DoRA í™œì„±í™”
)
````

````python
from datasets import load_dataset


# âœ… ë°ì´í„°ì…‹ ë¡œë“œ
dataset = load_dataset("jihye-moon/LawQA-Ko", split="train")

# âœ… EOS í† í°
EOS_TOKEN = tokenizer.eos_token

# âœ… Alpaca-style í”„ë¡¬í”„íŠ¸ í…œí”Œë¦¿
alpaca_prompt = """Below is an instruction that describes a task. Write a response that appropriately completes the request.

### Instruction:
{}

### Response:
{}"""


# âœ… Step 1: instruction / output í•„ë“œ ìƒì„±
def to_alpaca_format(example):
    return {
        "instruction": example["question"],
        "output": example["answer"]
    }

dataset = dataset.map(to_alpaca_format)


# âœ… Step 2: output ê¸¸ì´ í•„í„°ë§ (í† í° ê¸¸ì´ 516 ì´í•˜)
def is_output_short(example):
    tokenized = tokenizer(example["output"], truncation=False, add_special_tokens=False)
    return len(tokenized["input_ids"]) <= 516

filtered_dataset = dataset.filter(is_output_short)

# âœ… Step 3: text í•„ë“œ ìƒì„±
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    outputs = examples["output"]
    texts = []
    for instruction, output in zip(instructions, outputs):
        text = alpaca_prompt.format(instruction, output) + EOS_TOKEN
        texts.append(text)
    return { "text": texts }

formatted_dataset = filtered_dataset.map(formatting_prompts_func, batched=True)

# âœ… ê²°ê³¼ í™•ì¸
print(f"ğŸ“Š í•„í„°ë§ í›„ ìƒ˜í”Œ ìˆ˜: {len(formatted_dataset)}")
print(formatted_dataset[0])
````

````python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported  # BFloat16 ì§€ì› ì—¬ë¶€ í™•ì¸ í•¨ìˆ˜ ì„í¬íŠ¸


# SFTTrainerë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ í•™ìŠµ ì„¤ì •
# SFTTrainer ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
trainer = SFTTrainer(
    model = model,                           # í•™ìŠµí•  ëª¨ë¸
    tokenizer = tokenizer,                   # ì‚¬ìš©í•  í† í¬ë‚˜ì´ì €
    train_dataset = formatted_dataset,                 # í•™ìŠµí•  ë°ì´í„°ì…‹ â˜…â˜…â˜…â˜…â˜…â˜…â˜…â˜…
    dataset_text_field = "text",             # ë°ì´í„°ì…‹ì˜ í…ìŠ¤íŠ¸ í•„ë“œ ì´ë¦„ â˜…â˜…â˜…â˜…â˜…â˜…â˜…â˜…
    max_seq_length = 1024,         # ìµœëŒ€ ì‹œí€€ìŠ¤ ê¸¸ì´
    dataset_num_proc = 2,                    # ë°ì´í„°ì…‹ ì „ì²˜ë¦¬ì— ì‚¬ìš©í•  í”„ë¡œì„¸ìŠ¤ ìˆ˜ cpu
    packing = False,                         # ì§§ì€ ì‹œí€€ìŠ¤ì˜ ê²½ìš° packingì„ ë¹„í™œì„±í™” (í•™ìŠµ ì†ë„ 5ë°° í–¥ìƒ ê°€ëŠ¥)
    args = TrainingArguments(
        per_device_train_batch_size = 2,     # ë””ë°”ì´ìŠ¤ ë‹¹ ë°°ì¹˜ ì‚¬ì´ì¦ˆ
        gradient_accumulation_steps = 4,     # ê·¸ë˜ë””ì–¸íŠ¸ ëˆ„ì  ë‹¨ê³„ ìˆ˜
        warmup_steps = 5,                     # ì›Œë°ì—… ìŠ¤í… ìˆ˜
        # num_train_epochs = 1,               # ì „ì²´ í•™ìŠµ ì—í­ ìˆ˜ ì„¤ì • ê°€ëŠ¥
        max_steps = 60,                       # ìµœëŒ€ í•™ìŠµ ìŠ¤í… ìˆ˜
        learning_rate = 2e-4,                 # í•™ìŠµë¥ 
        fp16 = not is_bfloat16_supported(),   # BFloat16 ì§€ì› ì—¬ë¶€ì— ë”°ë¼ FP16 ì‚¬ìš©
        bf16 = is_bfloat16_supported(),       # BFloat16 ì‚¬ìš© ì—¬ë¶€
        logging_steps = 1,                    # ë¡œê¹… ë¹ˆë„
        optim = "adamw_8bit",                  # ì˜µí‹°ë§ˆì´ì € ì„¤ì • (8ë¹„íŠ¸ AdamW)
        weight_decay = 0.01,                  # ê°€ì¤‘ì¹˜ ê°ì‡ 
        lr_scheduler_type = "linear",         # í•™ìŠµë¥  ìŠ¤ì¼€ì¤„ëŸ¬ íƒ€ì…
        seed = 3407,                           # ëœë¤ ì‹œë“œ ì„¤ì •
        output_dir = "outputs",                # ì¶œë ¥ ë””ë ‰í† ë¦¬
        report_to = "none",
    ),
)
````

 Step 1. ë¬¸ì„œ ì¤€ë¹„ : ì§ˆë¬¸ + ë‹µë³€ í•˜ë‚˜ì˜ ë¬¸ì¥ìœ¼ë¡œ êµ¬ì„±

````python
from datasets import load_dataset

# ë°ì´í„°ì…‹ ë¡œë“œ
dataset = load_dataset("jihye-moon/LawQA-Ko", split="train")

# ë¬¸ì„œ ë¦¬ìŠ¤íŠ¸ ìƒì„±
documents = []

for example in dataset.select(range(1000)):  # ì˜ˆì‹œë¡œ 1000ê°œë§Œ ì‚¬ìš©
    question = example["question"].strip()
    answer = example["answer"].strip()
    full_text = f"{question}\n{answer}"  # ì§ˆë¬¸ê³¼ ë‹µë³€ ê²°í•©
    documents.append(full_text)

print(f"âœ… ë¬¸ì„œ ìˆ˜: {len(documents)}")
print("ğŸ“„ ì²« ë²ˆì§¸ ë¬¸ì„œ:\n", documents[0])
````

Step 2. ë¬¸ì„œ ì„ë² ë”©: KoSBERT ê¸°ë°˜
- sentence-transformersì˜ jhgan/ko-sbert-nli ëª¨ë¸ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

```
KoSBERTë€?
KoSBERTëŠ” ë¬¸ì¥ì„ ë²¡í„°ë¡œ ë³€í™˜í•˜ëŠ” í•œêµ­ì–´ íŠ¹í™” ì„ë² ë”© ëª¨ë¸ì…ë‹ˆë‹¤.

- SBERT(Sentence-BERT)ì˜ êµ¬ì¡°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•˜ë©°,
- í•œêµ­ì–´ ìì—°ì–´ ì¶”ë¡ (NLI) ë°ì´í„°ë¡œ íŒŒì¸íŠœë‹ ë˜ì–´ ìˆì–´ ë¬¸ì¥ ê°„ ì˜ë¯¸ ìœ ì‚¬ë„ ê³„ì‚°ì— íƒì›”í•©ë‹ˆë‹¤.
- ë³¸ ì‹¤ìŠµì—ì„œëŠ” sentence-transformers ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ í†µí•´ jhgan/ko-sbert-nli ëª¨ë¸ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.
- ì´ë¥¼ í†µí•´ "ì§ˆë¬¸ê³¼ ì˜ë¯¸ì ìœ¼ë¡œ ê°€ì¥ ë¹„ìŠ·í•œ ë¬¸ì„œ"ë¥¼ ë¹ ë¥´ê²Œ ì°¾ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
```

````python
from sentence_transformers import SentenceTransformer

# KoSBERT ëª¨ë¸ ë¡œë“œ
embedder = SentenceTransformer("jhgan/ko-sbert-nli")

# ë¬¸ì„œ ì„ë² ë”©
doc_embeddings = embedder.encode(documents, convert_to_numpy=True, show_progress_bar=True)
````

 Step 3. FAISSë¥¼ ì´ìš©í•œ ë²¡í„° ì¸ë±ìŠ¤ ìƒì„±
- FAISSëŠ” ê³ ì† ë²¡í„° ê²€ìƒ‰ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¡œ, ë¬¸ì„œ ê°„ ìœ ì‚¬ë„ ê¸°ë°˜ ê²€ìƒ‰ì— í™œìš©ë©ë‹ˆë‹¤.

````python
import faiss
import numpy as np

# ì„ë² ë”© ì°¨ì› í™•ì¸
dimension = doc_embeddings.shape[1]

# ë²¡í„° ì¸ë±ìŠ¤ ìƒì„±
index = faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)

print("âœ… FAISS ì¸ë±ìŠ¤ ìƒì„± ì™„ë£Œ!")
````

Step 4. ìœ ì‚¬ ë¬¸ì„œ ê²€ìƒ‰ í•¨ìˆ˜ ì •ì˜

````python
def search_documents(query, top_k=3):
    # ì¿¼ë¦¬ ì„ë² ë”©
    query_vec = embedder.encode([query])

    # ìœ ì‚¬ë„ ê¸°ë°˜ ê²€ìƒ‰
    D, I = index.search(np.array(query_vec), k=top_k)

    # ìƒìœ„ ë¬¸ì„œ ë°˜í™˜
    return [documents[i] for i in I[0]]
````

````python
# ì˜ˆì‹œ: ì§ˆë¬¸ì„ ì…ë ¥í•´ ê´€ë ¨ ë¬¸ì„œ ê²€ìƒ‰
query = "ë¶€ë™ì‚° ê³„ì•½ í•´ì§€ ì¡°ê±´ì€ ì–´ë–»ê²Œ ë˜ë‚˜ìš”?"
results = search_documents(query, top_k=3)

print("ğŸ§  ì‚¬ìš©ì ì§ˆë¬¸:", query)
print("\nğŸ“„ ê²€ìƒ‰ëœ ìœ ì‚¬ ë¬¸ì„œ:")
for i, doc in enumerate(results):
    print(f"\nğŸ”¹ [Top {i+1}]\n{doc[:300]}...")
````

Step 5. ê²€ìƒ‰ ê²°ê³¼ë¥¼ ê¸°ë°˜ìœ¼ë¡œ LLMì— ì§ˆë¬¸ í”„ë¡¬í”„íŠ¸ êµ¬ì„±

Dense Retrievalë¡œ ê°€ì ¸ì˜¨ ë¬¸ì„œë“¤ì„ LLM ì…ë ¥ í”„ë¡¬í”„íŠ¸ì— ì‚½ì…í•©ë‹ˆë‹¤.
ìš°ë¦¬ëŠ” Alpaca-style ëª¨ë¸ì„ ì‚¬ìš©í•˜ê³  ìˆìœ¼ë¯€ë¡œ, í”„ë¡¬í”„íŠ¸ëŠ” ì•„ë˜ì™€ ê°™ì€ í˜•íƒœë¡œ êµ¬ì„±í•©ë‹ˆë‹¤:

```
ë‹¤ìŒì€ ì°¸ê³ í•  ìˆ˜ ìˆëŠ” ë¬¸ì„œë“¤ì…ë‹ˆë‹¤:
{retrieved_doc_1}

---

{retrieved_doc_2}

---

...

### Instruction:
{user_query}

### Response:
```

````python
from unsloth import FastLanguageModel
from transformers import TextStreamer
import torch

# âœ… ëª¨ë¸ ë¡œë“œ
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "MLP-KTLim/llama-3-Korean-Bllossom-8B", # finetuning ì´ì „ ì›ë³¸ ëª¨ë¸ë¡œ ì‹¤ìŠµ
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True
)

# âœ… ì¶”ë¡  ìµœì í™”
FastLanguageModel.for_inference(model)
tokenizer.pad_token = tokenizer.eos_token
````

ì›ë³¸ì— ë°”ë¡œ ì§ˆë¬¸ í–ˆì„ ë•Œ ê²°ê³¼

````python
user_input = "ë¶€ë™ì‚° ê³„ì•½ í•´ì§€ ì¡°ê±´ì€ ì–´ë–»ê²Œ ë˜ë‚˜ìš”?"

# âœ… ì…ë ¥ì„ ëª¨ë¸ì´ ì²˜ë¦¬í•  ìˆ˜ ìˆëŠ” í…ì„œë¡œ ë³€í™˜
inputs = tokenizer(user_input, return_tensors="pt").to("cuda")
print(f"ğŸ“ ì§ˆë¬¸: {user_input}")

# âœ… ëª¨ë¸ ì‘ë‹µ ìƒì„± (ì¶”ë¡  ì˜µì…˜ ì ìš©)
outputs = model.generate(
    **inputs,
    max_new_tokens=256,        # ì‘ë‹µ ìµœëŒ€ ê¸¸ì´ ì œí•œ
    temperature=0.5,           # ì˜¨ë„ ì¡°ì ˆì„ í†µí•´ ì¼ê´€ëœ ì‘ë‹µ ìœ ë„
    top_p=0.85,                # ìƒìœ„ í™•ë¥  í† í°ë§Œ ìƒ˜í”Œë§
    repetition_penalty=1.2,    # ë°˜ë³µ ë°©ì§€
    do_sample=True             # ìƒ˜í”Œë§ ë°©ì‹ ì ìš©
)

# âœ… ì‘ë‹µ ë””ì½”ë”©
decoded_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("ğŸ¤– ì‘ë‹µ:", decoded_output)
````

Step 6. LLM í”„ë¡¬í”„íŠ¸ êµ¬ì„± ë° ì‘ë‹µ ìƒì„±

````python
def generate_answer_with_retrieval(query, top_k=3, max_context_len=1500):
    # ğŸ” ë¬¸ì„œ ê²€ìƒ‰
    retrieved_docs = search_documents(query, top_k=top_k)

    # ğŸ“„ ë¬¸ì„œ ë³‘í•© (ë„ˆë¬´ ê¸¸ë©´ ìë¦„)
    context = "\n\n---\n\n".join(retrieved_docs)
    if len(context) > max_context_len:
        context = context[:max_context_len]

    # ğŸ§¾ í”„ë¡¬í”„íŠ¸ êµ¬ì„± (Alpaca-style)
    prompt = (
        f"ë‹¤ìŒì€ ì°¸ê³ í•  ìˆ˜ ìˆëŠ” ë¬¸ì„œë“¤ì…ë‹ˆë‹¤:\n{context}\n\n"
        f"### Instruction:\n{query}\n\n"
        f"### Response:\n"
    )

    # ğŸ”¢ í† í¬ë‚˜ì´ì¦ˆ
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    # ğŸ¤– ì‘ë‹µ ìƒì„±
    outputs = model.generate(
        **inputs,
        max_new_tokens=256,
        temperature=0.7,
        top_p=0.9,
        repetition_penalty=1.1,
    )

    # ğŸ§¾ ë””ì½”ë”©
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    if "### Response:" in decoded:
        response = decoded.split("### Response:")[-1].strip()
    else:
        response = decoded.strip()

    return response
````

````python
# ìµœì¢… ì‹¤í–‰ ì˜ˆì‹œ

query = "ë¶€ë™ì‚° ê³„ì•½ í•´ì§€ ì¡°ê±´ì€ ì–´ë–»ê²Œ ë˜ë‚˜ìš”?"
response = generate_answer_with_retrieval(query)

print("ğŸ§  ì‚¬ìš©ì ì§ˆë¬¸:", query)
print("\nğŸ¤– LLM ì‘ë‹µ:\n", response)
````

### (2) Hybrid Search (BM25 + ë²¡í„° ê²€ìƒ‰)

Hybrid SearchëŠ” í‚¤ì›Œë“œ ê¸°ë°˜ì˜ BM25 ê²€ìƒ‰ê³¼ ì˜ë¯¸ ê¸°ë°˜ì˜ **ì„ë² ë”© ê²€ìƒ‰(Dense Retrieval)**ì„ ê²°í•©í•˜ì—¬ ì •í™•ë„ì™€ ë‹¤ì–‘ì„±ì„ ëª¨ë‘ í™•ë³´í•  ìˆ˜ ìˆëŠ” ê²€ìƒ‰ ê¸°ë²•ì…ë‹ˆë‹¤.

- Dense Retrievalì€ ì§ˆë¬¸ ì˜ë¯¸ì™€ ìœ ì‚¬í•œ ë¬¸ì„œë¥¼ ì°¾ì•„ë‚´ëŠ” ë° íš¨ê³¼ì ì…ë‹ˆë‹¤.
- ë‘ ê²€ìƒ‰ ê²°ê³¼ë¥¼ í†µí•©í•˜ì—¬ í‰ê·  ì ìˆ˜ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ìµœì¢… ìˆœìœ„ë¥¼ ì‚°ì¶œí•©ë‹ˆë‹¤.

BM25ë€?
BM25(Best Matching 25)ëŠ” ì •ë³´ ê²€ìƒ‰(IR)ì—ì„œ ê°€ì¥ ë„ë¦¬ ì“°ì´ëŠ” í‚¤ì›Œë“œ ê¸°ë°˜ ë¬¸ì„œ ìˆœìœ„í™” ì•Œê³ ë¦¬ì¦˜ì…ë‹ˆë‹¤.
ê²€ìƒ‰ì–´ì™€ ë¬¸ì„œ ì‚¬ì´ì˜ ë‹¨ì–´ ë¹ˆë„(TF)ì™€ ì—­ë¬¸ì„œ ë¹ˆë„(IDF)ë¥¼ ë°”íƒ•ìœ¼ë¡œ, ê´€ë ¨ë„ ì ìˆ˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

- ë‹¨ìˆœí•œ ë‹¨ì–´ ë§¤ì¹­ì´ì§€ë§Œ, ì§ˆì˜ì–´ì™€ ë¬¸ì„œ ê°„ì˜ í…ìŠ¤íŠ¸ì  ì¼ì¹˜ë¥¼ ì˜ ë°˜ì˜í•©ë‹ˆë‹¤.

BM25ëŠ” ì¼ë°˜ì ìœ¼ë¡œ Whoosh, Elasticsearch, Lucene, scikit-learn TfidfVectorizer ë“±ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

Step 1. ë¬¸ì„œ ì¤€ë¹„ (ì• ë‹¨ê³„ ë™ì¼)

````python
from datasets import load_dataset

dataset = load_dataset("jihye-moon/LawQA-Ko", split="train")

documents = []
for example in dataset.select(range(1000)):
    question = example["question"].strip()
    answer = example["answer"].strip()
    full_text = f"{question}\n{answer}"
    documents.append(full_text)
````

Step 2. ì„ë² ë”© ë²¡í„° & TF-IDF ë²¡í„° ì¤€ë¹„

````python
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import faiss

# KoSBERT ì„ë² ë”©
embedder = SentenceTransformer("jhgan/ko-sbert-nli")
doc_embeddings = embedder.encode(documents, convert_to_numpy=True, show_progress_bar=True)

# TF-IDF ë²¡í„°í™”
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# FAISS ì¸ë±ìŠ¤ ìƒì„±
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)
````

Step 3. Hybrid ê²€ìƒ‰ í•¨ìˆ˜ ì •ì˜

````python
def hybrid_search(query, top_k=3):
    # 1. TF-IDF (BM25 ìœ ì‚¬ë„)
    query_tfidf = vectorizer.transform([query])
    bm25_scores = cosine_similarity(query_tfidf, tfidf_matrix).flatten()

    # 2. Dense ì„ë² ë”© ìœ ì‚¬ë„
    query_dense = embedder.encode([query])
    D, I_dense = index.search(np.array(query_dense), k=top_k * 2)  # í›„ë³´ ë„‰ë„‰íˆ

    # 3. Hybrid ì ìˆ˜ ê³„ì‚°
    hybrid_scores = {}
    dense_indices = I_dense[0]
    bm25_top = np.argsort(bm25_scores)[-top_k*2:]

    for i in set(dense_indices).union(set(bm25_top)):
        bm25_score = bm25_scores[i]
        dense_score = 1 - np.linalg.norm(query_dense - doc_embeddings[i])
        hybrid_scores[i] = (bm25_score + dense_score) / 2

    # 4. ìƒìœ„ ë¬¸ì„œ ë°˜í™˜
    sorted_docs = sorted(hybrid_scores.items(), key=lambda x: x[1], reverse=True)[:top_k]
    return [documents[i] for i, _ in sorted_docs]
````

Step 5. Hybrid ê²€ìƒ‰ ê¸°ë°˜ ì‘ë‹µ í•¨ìˆ˜ ì •ì˜

````python
def generate_hybrid_answer(query, top_k=3, max_context_len=1500):
    retrieved_docs = hybrid_search(query, top_k=top_k)
    context = "\n\n---\n\n".join(retrieved_docs)
    if len(context) > max_context_len:
        context = context[:max_context_len]

    prompt = (
        f"ë‹¤ìŒì€ ì°¸ê³ í•  ìˆ˜ ìˆëŠ” ë¬¸ì„œë“¤ì…ë‹ˆë‹¤:\n{context}\n\n"
        f"### Instruction:\n{query}\n\n"
        f"### Response:\n"
    )

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(
        **inputs,
        max_new_tokens=200,
        temperature=0.7,
        top_p=0.9,
        repetition_penalty=1.1,
    )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    if "### Response:" in decoded:
        response = decoded.split("### Response:")[-1].strip()
    else:
        response = decoded.strip()
    return response
````

### (3) Rerankingì„ í†µí•œ ê²€ìƒ‰ ê²°ê³¼ ê°œì„

Step 1. Hybrid Search ê¸°ë°˜ ì´ˆê¸° í›„ë³´ ë¬¸ì„œ 10ê°œ ì¶”ì¶œ

````python
# ì‚¬ìš©ì ì§ˆë¬¸
query = "ë¶€ë™ì‚° ê³„ì•½ í•´ì§€ ì¡°ê±´ì€ ì–´ë–»ê²Œ ë˜ë‚˜ìš”?"

# 1. TF-IDF ìœ ì‚¬ë„ ê³„ì‚°
query_tfidf = vectorizer.transform([query])
bm25_scores = cosine_similarity(query_tfidf, tfidf_matrix).flatten()
bm25_top = np.argsort(bm25_scores)[::-1][:10]

# 2. Dense ì„ë² ë”© ìœ ì‚¬ë„ ê³„ì‚°
query_dense = embedder.encode([query])
D, I_dense = index.search(np.array(query_dense), k=10)
dense_top = I_dense[0]

# 3. Hybrid Score ê³„ì‚° (BM25 + Dense í‰ê· )
hybrid_scores = {}
for i in set(bm25_top).union(set(dense_top)):
    bm25_score = bm25_scores[i]
    dense_score = 1 - np.linalg.norm(query_dense - doc_embeddings[i])
    hybrid_scores[i] = (bm25_score + dense_score) / 2

# í›„ë³´ ë¬¸ì„œ ì¶”ì¶œ (10ê°œ)
top_candidates = sorted(hybrid_scores.items(), key=lambda x: x[1], reverse=True)[:10]
candidate_docs = [(query, documents[idx]) for idx, _ in top_candidates]
````

````python
from unsloth import FastLanguageModel
import torch

# ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸° (íŒŒì¸íŠœë‹ ëª¨ë¸ ë˜ëŠ” ì›ë³¸)
model_path = "Dora_model"  # or "Dora_model"
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_path,
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)

FastLanguageModel.for_inference(model)
tokenizer.pad_token = tokenizer.eos_token
````

Step 2. RAG ì‘ë‹µ ìƒì„± í•¨ìˆ˜ ì •ì˜

````python
def generate_answer_with_context(query, context_docs, max_context_len=1500):
    context = "\n\n---\n\n".join(context_docs)
    if len(context) > max_context_len:
        context = context[:max_context_len]

    prompt = (
        f"ë‹¤ìŒì€ ì°¸ê³ í•  ìˆ˜ ìˆëŠ” ë¬¸ì„œë“¤ì…ë‹ˆë‹¤:\n{context}\n\n"
        f"### Instruction:\n{query}\n\n"
        f"### Response:\n"
    )
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(
        **inputs,
        max_new_tokens=256,
        temperature=0.7,
        top_p=0.9,
        repetition_penalty=1.1,
    )
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return decoded.split("### Response:")[-1].strip()
````
